```python
# ==============================================================================
# 0. IMPORTS AND INITIAL SETUP
# ==============================================================================
print("\n" + "="*80)
print("... 0. SCRIPT INITIALIZED FOR FROZEN ESMC EXPERIMENT ...")
print("    Training with a frozen ESMC backbone and a trainable projection head.")

# ==============================================================================
# 1. IMPORTS AND INITIAL SETUP
# ==============================================================================
import os
import sys
import time
import json
import random
import pickle
from pathlib import Path
from datetime import datetime

# Standard data science libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings

# Machine learning libraries
from sklearn.model_selection import KFold, GroupKFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr

# PyTorch
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW

# --- MODIFIED: Imports for ESMC ---
from esm.models.esmc import ESMC
from esm.sdk.api import ESMProtein, LogitsConfig

from typing import List

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
warnings.filterwarnings('ignore', category=UserWarning)

# --- MODIFIED: Title reflects the use of a frozen ESMC model ---
print("🧬 BALM Adaptation with a Frozen ESMC Backbone")
print("⚡ Fast Training with 5-Fold Cross-Validation and Pre-computed Embeddings")
print("="*80)

def setup_reproducibility(seed=42):
    """Setup complete reproducibility."""
    print(f"🔒 Setting up reproducibility with seed {seed}")
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)
    print(f"✅ Reproducibility setup complete")

setup_reproducibility(42)

# ==============================================================================
# 2. ESMC EMBEDDING GENERATOR (MODIFIED for ESMC)
# ==============================================================================

class ESMCEmbeddingGenerator:
    """
    A class to load a frozen ESMC model and generate embeddings for a list of sequences.
    This is used for pre-computation.
    """
    def __init__(
        self,
        model_name="esmc_300m",
        device="auto"
    ):
        self.model_name = model_name

        if device == "auto":
            self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        else:
            self.device = torch.device(device)

        print(f"🔧 Initializing ESMCEmbeddingGenerator on {self.device}")

        self.dtype = torch.float16 if self.device.type == 'cuda' else torch.float32
        print(f"📥 Loading ESMC model: {model_name} with dtype: {self.dtype}")

        self.model = ESMC.from_pretrained(model_name).to(self.device)
        self.model.eval()

        print("   ⚠️ All base model weights are frozen for embedding generation.")
        for param in self.model.parameters():
            param.requires_grad = False

        # Determine embedding size by processing a dummy sequence
        with torch.no_grad():
            dummy_protein = ESMProtein(sequence="A")
            dummy_tensor = self.model.encode(dummy_protein)
            dummy_output = self.model.logits(
                dummy_tensor, LogitsConfig(sequence=True, return_embeddings=True)
            )
            self.embedding_size = dummy_output.embeddings.shape[-1]

        print(f"   ✅ Model loaded. Embedding size from dummy run: {self.embedding_size}")

    @torch.no_grad()
    def get_embeddings(self, sequences: List[str]):
        all_embeddings = []
        for seq in sequences:
            # CRITICAL: Replace '|' with '<s><s>' for multi-chain support
            processed_seq = seq.replace('|', '<s><s>')

            protein = ESMProtein(sequence=processed_seq)
            protein_tensor = self.model.encode(protein)
            logits_output = self.model.logits(
                protein_tensor, LogitsConfig(sequence=True, return_embeddings=True)
            )
            # Mean pooling over the full sequence (including internal separator tokens)
            mean_embedding = logits_output.embeddings.mean(dim=1).squeeze(0)
            all_embeddings.append(mean_embedding)

        return torch.stack(all_embeddings).to(torch.float32).cpu()


# ==============================================================================
# 2.5. PRE-COMPUTE AND CACHE EMBEDDINGS (MODIFIED for ESMC)
# ==============================================================================
print("\n" + "="*80)
print("... 2.5. PRE-COMPUTING AND CACHING EMBEDDINGS ...")

def generate_and_cache_embeddings(df: pd.DataFrame, cache_path="esmc_embeddings.pkl", batch_size=16):
    """
    Generates embeddings for all unique protein sequences and saves them to a cache file.
    If the cache file exists, it loads the embeddings from there.
    """
    if os.path.exists(cache_path):
        print(f"📂 Loading pre-computed embeddings from '{cache_path}'...")
        with open(cache_path, 'rb') as f:
            embedding_dict, embedding_size = pickle.load(f)
        print(f"✅ Loaded {len(embedding_dict)} embeddings. Embedding size: {embedding_size}")
        return embedding_dict, embedding_size

    print("🛠️ No cache found. Generating new embeddings...")

    unique_seqs = pd.concat([df['Target'], df['proteina']]).unique()
    print(f"Found {len(unique_seqs)} unique protein sequences to process.")

    generator = ESMCEmbeddingGenerator()
    embedding_size = generator.embedding_size

    embedding_dict = {}

    for i in tqdm(range(0, len(unique_seqs), batch_size), desc="Generating Embeddings"):
        batch_seqs = unique_seqs[i:i+batch_size]
        batch_embeddings = generator.get_embeddings(list(batch_seqs))

        for seq, emb in zip(batch_seqs, batch_embeddings):
            embedding_dict[seq] = emb

    print(f"💾 Saving {len(embedding_dict)} embeddings to '{cache_path}'...")
    with open(cache_path, 'wb') as f:
        pickle.dump((embedding_dict, embedding_size), f)

    del generator
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    print("✅ Embeddings generated and cached successfully.")
    return embedding_dict, embedding_size

# ==============================================================================
# 3. LOAD YOUR DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 3. LOADING DATA FROM Data.csv ...")

protein1_col = 'Target'
protein2_col = 'proteina'
target_col = 'Y'
pdb_col_raw = 'PDB'
subgroup_col = 'Subgroup'
source_dataset_col = 'Source Data Set' # NEW

try:
    # IMPORTANT: Update this path to your actual file location
    df_raw = pd.read_csv(r'/teamspace/studios/this_studio/PPB_Affinity_Sequences_Final (version 1).csv')
    print(f"Successfully loaded Data.csv with {len(df_raw)} rows.")
    df = df_raw[[protein1_col, protein2_col, target_col, pdb_col_raw,subgroup_col,source_dataset_col]].copy()
    df.rename(columns={
        protein1_col: 'Target',
        protein2_col: 'proteina',
        target_col: 'Y',
        pdb_col_raw: 'PDB',
        subgroup_col : 'Subgroup',
        source_dataset_col: 'SourceDataSet' # NEW
    }, inplace=True)

    print("\n--- Initial Data Check ---")
    print(df.head())
    missing_values = df.isnull().sum()
    print("\nMissing values per column:\n", missing_values)

    if missing_values.sum() > 0:
        df.dropna(subset=['Target', 'proteina', 'Y', 'PDB'], inplace=True)
        print(f"Dataset now has {len(df)} rows after cleaning.")

    df['Target'] = df['Target'].astype(str)
    df['proteina'] = df['proteina'].astype(str)
    df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
    df['PDB'] = df['PDB'].astype(str)
    df['Subgroup'] = df['Subgroup'].astype(str)
    df['SourceDataSet'] = df['SourceDataSet'].astype(str) # NEW

    df.dropna(subset=['Y'], inplace=True)

    print(f"\n✅ Data is loaded and cleaned. Final dataset size: {len(df)} samples.")
    df.info()

except FileNotFoundError:
    print("❌ ERROR: 'Data.csv' not found. Please ensure the path is correct.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB','Subgroup'])
except KeyError as e:
    print(f"❌ ERROR: Column {e} not found in 'Data.csv'. Please check column names.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB','Subgroup'])

# ==============================================================================
# 4. EXPERIMENT CONFIGURATION
# ==============================================================================
print("\n" + "="*80)
print("... 4. CONFIGURING THE EXPERIMENT ...")

CV_STRATEGY = 'GroupKFold'
N_SPLITS = 5
EPOCHS = 30
PATIENCE = 15
BATCH_SIZE = 16
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 256
DROPOUT = 0.1

# Output directory for results
RESULTS_DIR = "cv_results_frozen_esmc"
os.makedirs(RESULTS_DIR, exist_ok=True)
print(f"Results will be saved in: {RESULTS_DIR}")

print(f"✅ Configuration set. Using CV Strategy: {CV_STRATEGY}")
print(f"   Epochs: {EPOCHS}, Patience: {PATIENCE}")
print(f"   Training with a frozen ESMC backbone. Only the projection head will be trained.")

# ==============================================================================
# 5. ADAPTED MODEL AND DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 5. DEFINING ADAPTED MODEL AND DATASET ...")

class BALMProjectionHead(nn.Module):
    def __init__(self, embedding_size, projected_size, projected_dropout):
        super().__init__()
        self.protein_projection = nn.Linear(embedding_size, projected_size)
        self.proteina_projection = nn.Linear(embedding_size, projected_size)
        self.dropout = nn.Dropout(projected_dropout)
        self.loss_fn = nn.MSELoss()
        print(f"✅ BALMProjectionHead initialized with embedding_size: {embedding_size}")

    def forward(self, protein_embedding, proteina_embedding, labels=None):
        protein_embedding = self.dropout(protein_embedding)
        proteina_embedding = self.dropout(proteina_embedding)
        protein_projected = self.protein_projection(protein_embedding)
        proteina_projected = self.proteina_projection(proteina_embedding)
        protein_projected = F.normalize(protein_projected, p=2, dim=1)
        proteina_projected = F.normalize(proteina_projected, p=2, dim=1)
        cosine_similarity = F.cosine_similarity(protein_projected, proteina_projected)
        cosine_similarity = torch.clamp(cosine_similarity, -0.9999, 0.9999)
        output = {"cosine_similarity": cosine_similarity}
        if labels is not None:
            loss = self.loss_fn(cosine_similarity, labels)
            output["loss"] = loss
        return output

class BALMForRegression(nn.Module):
    def __init__(self, embedding_size, projected_size, projected_dropout, pkd_bounds):
        super().__init__()
        self.projection_head = BALMProjectionHead(embedding_size, projected_size, projected_dropout)
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def forward(self, batch_input):
        protein_emb = batch_input["protein_embedding"]
        proteina_emb = batch_input["proteina_embedding"]
        labels = batch_input.get("labels")

        proj_output = self.projection_head(protein_emb, proteina_emb, labels)

        # Return a unified dictionary
        output = {
            "cosine_similarity": proj_output["cosine_similarity"],
            "original_pkds": batch_input["original_pkds"],
            "pdb_groups": batch_input["pdb_groups"],
            "subgroups": batch_input["subgroups"],
            "source_dataset": batch_input["source_dataset"]
        }
        if "loss" in proj_output:
            output["loss"] = proj_output["loss"]
        return output

class ProteinPairEmbeddingDataset(Dataset):
    def __init__(self, dataframe, embedding_dict, pkd_bounds):
        self.data = dataframe.reset_index(drop=True)
        self.embedding_dict = embedding_dict
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        original_pkd = float(row['Y'])
        scaled_label = ((original_pkd - self.pkd_lower) / self.pkd_range) * 2 - 1

        protein_seq = row["Target"]
        proteina_seq = row["proteina"]

        return {
            "protein_embedding": self.embedding_dict[protein_seq],
            "proteina_embedding": self.embedding_dict[proteina_seq],
            "labels": torch.tensor(scaled_label, dtype=torch.float32),
            "original_pkds": torch.tensor(original_pkd, dtype=torch.float32),
            "pdb_groups": row["PDB"],
            "subgroups": row["Subgroup"],
            "source_dataset": row["SourceDataSet"] # NEW
        }

def collate_fn_embeddings(batch):
    return {
        "protein_embedding": torch.stack([item['protein_embedding'] for item in batch]),
        "proteina_embedding": torch.stack([item['proteina_embedding'] for item in batch]),
        "labels": torch.stack([item['labels'] for item in batch]),
        "original_pkds": torch.stack([item['original_pkds'] for item in batch]),
        "pdb_groups": [item['pdb_groups'] for item in batch],
        "subgroups": [item['subgroups'] for item in batch],
        "source_dataset": [item['source_dataset'] for item in batch] # NEW
    }

# ==============================================================================
# 6. METRICS AND EVALUATION FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6. DEFINING METRICS AND EVALUATION ...")

def concordance_index(y_true, y_pred):
    """Calculates the Concordance Index (C-index) for regression."""
    try:
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        n = len(y_true)
        if n < 2: return 1.0
        concordant, discordant, tied_pred = 0, 0, 0
        for i in range(n):
            for j in range(i + 1, n):
                if y_true[i] == y_true[j]: continue
                if y_pred[i] == y_pred[j]:
                    tied_pred += 1
                    continue
                if (y_pred[i] > y_pred[j] and y_true[i] > y_true[j]) or \
                   (y_pred[i] < y_pred[j] and y_true[i] < y_true[j]):
                    concordant += 1
                else:
                    discordant += 1
        total_pairs = concordant + discordant + tied_pred
        return 1.0 if total_pairs == 0 else (concordant + 0.5 * tied_pred) / total_pairs
    except Exception:
        return np.nan

def evaluate_model(model, data_loader, device, pkd_bounds):
    model.eval()
    all_labels, all_preds_pkd, all_pdb_groups, all_subgroups, all_sources = [], [], [], [], []
    pkd_lower, pkd_range = pkd_bounds[0], pkd_bounds[1] - pkd_bounds[0]

    with torch.no_grad():
        for batch in data_loader:
            # Move Tensors to GPU, keep metadata lists on CPU
            batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}

            outputs = model(batch_gpu)
            cosine_sim = outputs['cosine_similarity'].cpu()
            # Inverse scaling
            preds_pkd = ((cosine_sim + 1) / 2) * pkd_range + pkd_lower

            all_labels.extend(batch['original_pkds'].numpy())
            all_preds_pkd.extend(preds_pkd.numpy())
            all_pdb_groups.extend(batch['pdb_groups'])
            all_subgroups.extend(batch['subgroups'])
            all_sources.extend(batch['source_dataset'])

    # Standardize names to match the return statement
    y_true, y_pred = np.array(all_labels), np.array(all_preds_pkd)

    mask = ~np.isnan(y_true) & ~np.isnan(y_pred)
    labels_masked, preds_masked = y_true[mask], y_pred[mask]

    if len(labels_masked) < 2:
        metrics = {"rmse": np.nan, "pearson": np.nan, "spearman": np.nan, "ci": np.nan}
    else:
        metrics = {
            'rmse': np.sqrt(mean_squared_error(labels_masked, preds_masked)),
            'pearson': pearsonr(labels_masked, preds_masked)[0],
            'spearman': spearmanr(labels_masked, preds_masked)[0],
            'ci': concordance_index(labels_masked, preds_masked)
        }

    # Return exactly 6 items
    return metrics, y_true, y_pred, np.array(all_pdb_groups), np.array(all_subgroups), np.array(all_sources)

def plot_regression(y_true, y_pred, metrics, title, filename=None):
    plt.figure(figsize=(8, 8))
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, edgecolor='k', s=50)
    min_val, max_val = min(np.min(y_true), np.min(y_pred)), max(np.max(y_true), np.max(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Identity Line (y=x)')
    metrics_text = f"RMSE: {metrics['rmse']:.4f}\nPearson: {metrics['pearson']:.4f}\nSpearman: {metrics['spearman']:.4f}\nCI: {metrics['ci']:.4f}"
    plt.text(0.05, 0.95, metrics_text, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))
    plt.title(title, fontsize=16)
    plt.xlabel("Actual pKd", fontsize=14)
    plt.ylabel("Predicted pKd", fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    if filename:
        plt.savefig(filename)
        print(f"Plot saved to {filename}")
    plt.show()

# ==============================================================================
# 7. CROSS-VALIDATION TRAINING (MODIFIED for Frozen ESMC)
# ==============================================================================
print("\n" + "="*80)
print("... 7. RUNNING 'COLD TARGET' (GROUPKFOLD) STRATEGY WITH FROZEN ESMC ...")

def get_groupkfold_splits(df: pd.DataFrame, n_splits: int):
    print(f"\n❄️ Preparing {n_splits}-Fold **GroupKFold (Cold Target)** splits...")
    gkf = GroupKFold(n_splits=n_splits)
    groups = df['PDB'].factorize()[0]
    for fold, (train_idx, test_idx) in enumerate(gkf.split(df, groups=groups)):
        train_targets = set(df.iloc[train_idx]['PDB'].unique())
        test_targets = set(df.iloc[test_idx]['PDB'].unique())
        if not train_targets.intersection(test_targets):
            print(f"  ✅ Fold {fold+1}: No 'PDB' group leakage detected.")
        else:
            print(f"  ❌ Fold {fold+1} FAILED: PDB group leakage detected!")
        yield train_idx, test_idx

def run_cold_target_cv_frozen_esmc():
    if df.empty:
        print("⚠️ Skipping training due to empty dataframe.")
        return

    # --- MODIFIED: Use the correct cache path for ESMC embeddings ---
    protein_embeddings, embedding_size = generate_and_cache_embeddings(df, cache_path="esmc_embeddings.pkl")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    PKD_LOWER_BOUND, PKD_UPPER_BOUND = df['Y'].min(), df['Y'].max()
    PKD_BOUNDS = (PKD_LOWER_BOUND, PKD_UPPER_BOUND)
    print(f"Global pKd bounds for scaling: [{PKD_LOWER_BOUND:.2f}, {PKD_UPPER_BOUND:.2f}]")

    print(f"\n🚀 Starting {N_SPLITS}-Fold **{CV_STRATEGY}** Cross-Validation with Frozen ESMC backbone...")

    splits = get_groupkfold_splits(df, N_SPLITS)
    fold_results, all_folds_y_true, all_folds_y_pred = [], [], []

    for fold, (train_ids, val_ids) in enumerate(splits):
        fold_num = fold + 1
        print(f"\n===== STRATEGY: {CV_STRATEGY} | FOLD {fold_num}/{N_SPLITS} =====")

        train_df, val_df = df.iloc[train_ids], df.iloc[val_ids]

        train_dataset = ProteinPairEmbeddingDataset(train_df, protein_embeddings, PKD_BOUNDS)
        val_dataset = ProteinPairEmbeddingDataset(val_df, protein_embeddings, PKD_BOUNDS)
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn_embeddings)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE*2, shuffle=False, collate_fn=collate_fn_embeddings)

        model = BALMForRegression(
            embedding_size=embedding_size,
            projected_size=PROJECTED_SIZE,
            projected_dropout=DROPOUT,
            pkd_bounds=PKD_BOUNDS
        ).to(device)

        optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)
        print(f"   Optimizing {sum(p.numel() for p in model.parameters() if p.requires_grad)} trainable parameters.")

        best_val_metric = float('inf')
        patience_counter = 0
        fold_model_path = os.path.join(RESULTS_DIR, f"best_model_fold_{fold_num}.pth")

        for epoch in range(EPOCHS):
            model.train()
            total_train_loss = 0.0
            progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} (Fold {fold_num})", leave=False)

            for batch in progress_bar:
                optimizer.zero_grad()
                batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
                batch_gpu['pdb_groups'] = batch['pdb_groups']
                batch_gpu['subgroups'] = batch['subgroups']

                outputs = model(batch_gpu)
                loss = outputs['loss']
                loss.backward()
                optimizer.step()
                total_train_loss += loss.item()
                progress_bar.set_postfix({'batch_loss': f"{loss.item():.4f}"})

            avg_train_loss = total_train_loss / len(train_loader)

            val_metrics, _, _, _, _, _ = evaluate_model(model, val_loader, device, PKD_BOUNDS)

            current_val_metric = val_metrics['rmse']

            print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {avg_train_loss:.4f} | Val RMSE: {current_val_metric:.4f} | Val Pearson: {val_metrics['pearson']:.4f}")

            if current_val_metric < best_val_metric:
                best_val_metric = current_val_metric
                patience_counter = 0
                torch.save(model.state_dict(), fold_model_path)
                print(f"   -> New best model for Fold {fold_num} saved with Val RMSE: {best_val_metric:.4f}")
            else:
                patience_counter += 1

            if patience_counter >= PATIENCE:
                print(f"⏳ Early stopping triggered at epoch {epoch+1} for Fold {fold_num}.")
                break

        print(f"Fold {fold_num} training complete.")

        if os.path.exists(fold_model_path):
            model.load_state_dict(torch.load(fold_model_path))

        final_val_metrics, y_true_fold, y_pred_fold, pdb_groups_fold, subgroups_fold, sources_fold = evaluate_model(model, val_loader, device, PKD_BOUNDS)

        print(f"Fold {fold_num} Final Validation Metrics:")
        for key, value in final_val_metrics.items(): print(f"  - {key.upper()}: {value:.4f}")

        fold_results.append(final_val_metrics)
        all_folds_y_true.extend(y_true_fold)
        all_folds_y_pred.extend(y_pred_fold)

        fold_output_df = pd.DataFrame({
            'label': y_true_fold,
            'prediction': y_pred_fold,
            'residual': y_true_fold - y_pred_fold,
            'abs_residual': np.abs(y_true_fold - y_pred_fold),
            'PDB': pdb_groups_fold,
            'Subgroup': subgroups_fold,
            'Source Data Set': sources_fold # Ensure this is included
        })
        fold_output_filename = os.path.join(RESULTS_DIR, f"fold_{fold_num}_predictions.csv")
        fold_output_df.to_csv(fold_output_filename, index=False)
        print(f"Predictions for Fold {fold_num} saved to {fold_output_filename}")

    print(f"\n📈 Generating combined regression plot for {CV_STRATEGY}")
    combined_y_true, combined_y_pred = np.array(all_folds_y_true), np.array(all_folds_y_pred)
    overall_metrics = {
        'rmse': np.sqrt(mean_squared_error(combined_y_true, combined_y_pred)),
        'pearson': pearsonr(combined_y_true, combined_y_pred)[0],
        'spearman': spearmanr(combined_y_true, combined_y_pred)[0],
        'ci': concordance_index(combined_y_true, combined_y_pred)
    }

    plot_regression(
        combined_y_true, combined_y_pred, overall_metrics,
        title=f"Overall Regression for {CV_STRATEGY} ({N_SPLITS}-Fold CV) with Frozen ESMC",
        filename=os.path.join(RESULTS_DIR, "overall_regression_plot.png")
    )

    print(f"\n✅ Final Results Summary for {CV_STRATEGY} with Frozen ESMC:")
    results_df = pd.DataFrame(fold_results)
    summary_df = pd.DataFrame({'Mean': results_df.mean(), 'Std Dev': results_df.std()})
    print(summary_df)

    summary_filename = os.path.join(RESULTS_DIR, "cv_summary_metrics.csv")
    summary_df.to_csv(summary_filename)
    print(f"Overall CV summary metrics saved to {summary_filename}")


# --- Call the CV function ---
run_cold_target_cv_frozen_esmc()

# ==============================================================================
# 8. FINAL NOTES AND RECOMMENDATIONS
# ==============================================================================
print("\n" + "="*80)
print("... 8. FINAL NOTES AND RECOMMENDATIONS ...")
print("""
🎉 CONGRATULATIONS! You've successfully implemented BALM using pre-computed embeddings
from a **Frozen ESMC Backbone**!

📊 KEY ADVANTAGES OF THIS APPROACH:
• **Extreme Speed:** Training is incredibly fast because the large ESMC model is not used in the training loop.
• **Memory Efficiency:** Requires very little GPU memory during training, allowing for much larger batch sizes.
• **Simplicity:** The model architecture and training pipeline are simpler compared to fine-tuning approaches.
• **State-of-the-Art Representations:** You are now using embeddings from the ESMC family of models, which are specialized for protein representation learning.

📉 POTENTIAL LIMITATIONS:
• **Static Features:** The ESMC embeddings are fixed and do not adapt to your specific dataset during training. This might limit the model's performance on tasks that are very different from ESMC's pre-training objectives.
• **Less Flexibility:** Unlike fine-tuning, this approach cannot learn new sequence-level patterns; it can only learn how to combine the existing fixed features.

🚀 PERFORMANCE OPTIMIZATION TIPS:
1.  **Batch Size:** Since this approach is memory-efficient, you can significantly increase the `BATCH_SIZE` to speed up training and potentially get more stable gradient updates.
2.  **Projection Head Architecture:** The bottleneck is now the small projection head. You could experiment with a deeper or wider head (e.g., adding more linear layers and non-linearities) to see if it improves performance.
3.  **Learning Rate:** The optimal learning rate for training just the head might be different from the one used for fine-tuning. Experiment with values from 1e-4 to 1e-3.

🎯 NEXT STEPS:
1.  **Compare Results:** Directly compare the performance metrics (RMSE, Pearson, etc.) from this run with your previous ESM-2 or LoRA fine-tuning runs. This will quantify the benefit of using ESMC embeddings for your specific task.
2.  **Experiment with Different ESMC Models:** You can easily swap the `model_name` in the `ESMCEmbeddingGenerator` to a larger ESMC model (e.g., 'esmc_600m') and regenerate the embeddings to see how the base model size impacts performance.
""")
print("\n✅ Frozen ESMC experiment complete! Happy protein modeling! 🧬🤖")
```

```python
# ==============================================================================
# 0. IMPORTS AND INITIAL SETUP
# ==============================================================================
print("\n" + "="*80)
print("... 0. SCRIPT INITIALIZED FOR FROZEN ESMC EXPERIMENT ...")
print("    Training with a frozen ESMC backbone and a trainable projection head.")

# ==============================================================================
# 1. IMPORTS AND INITIAL SETUP
# ==============================================================================
import os
import sys
import time
import json
import random
import pickle
from pathlib import Path
from datetime import datetime

# Standard data science libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings

# Machine learning libraries
from sklearn.model_selection import KFold, GroupKFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr

# PyTorch
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW

# --- MODIFIED: Imports for ESMC ---
from esm.models.esmc import ESMC
from esm.sdk.api import ESMProtein, LogitsConfig

from typing import List

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
warnings.filterwarnings('ignore', category=UserWarning)

# --- MODIFIED: Title reflects the use of a frozen ESMC model ---
print("🧬 BALM Adaptation with a Frozen ESMC Backbone")
print("⚡ Fast Training with 5-Fold Cross-Validation and Pre-computed Embeddings")
print("="*80)

def setup_reproducibility(seed=42):
    """Setup complete reproducibility."""
    print(f"🔒 Setting up reproducibility with seed {seed}")
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)
    print(f"✅ Reproducibility setup complete")

setup_reproducibility(42)

# ==============================================================================
# 2. ESMC EMBEDDING GENERATOR (MODIFIED for ESMC)
# ==============================================================================

class ESMCEmbeddingGenerator:
    """
    A class to load a frozen ESMC model and generate embeddings for a list of sequences.
    This is used for pre-computation.
    """
    def __init__(
        self,
        model_name="esmc_300m",
        device="auto"
    ):
        self.model_name = model_name

        if device == "auto":
            self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        else:
            self.device = torch.device(device)

        print(f"🔧 Initializing ESMCEmbeddingGenerator on {self.device}")

        self.dtype = torch.float16 if self.device.type == 'cuda' else torch.float32
        print(f"📥 Loading ESMC model: {model_name} with dtype: {self.dtype}")

        self.model = ESMC.from_pretrained(model_name).to(self.device)
        self.model.eval()

        print("   ⚠️ All base model weights are frozen for embedding generation.")
        for param in self.model.parameters():
            param.requires_grad = False

        # Determine embedding size by processing a dummy sequence
        with torch.no_grad():
            dummy_protein = ESMProtein(sequence="A")
            dummy_tensor = self.model.encode(dummy_protein)
            dummy_output = self.model.logits(
                dummy_tensor, LogitsConfig(sequence=True, return_embeddings=True)
            )
            self.embedding_size = dummy_output.embeddings.shape[-1]

        print(f"   ✅ Model loaded. Embedding size from dummy run: {self.embedding_size}")

    @torch.no_grad()
    def get_embeddings(self, sequences: List[str]):
        all_embeddings = []
        for seq in sequences:
            # CRITICAL: Replace '|' with '<s><s>' for multi-chain support
            processed_seq = seq.replace('|', '<s><s>')

            protein = ESMProtein(sequence=processed_seq)
            protein_tensor = self.model.encode(protein)
            logits_output = self.model.logits(
                protein_tensor, LogitsConfig(sequence=True, return_embeddings=True)
            )
            # Mean pooling over the full sequence (including internal separator tokens)
            mean_embedding = logits_output.embeddings.mean(dim=1).squeeze(0)
            all_embeddings.append(mean_embedding)

        return torch.stack(all_embeddings).to(torch.float32).cpu()


# ==============================================================================
# 2.5. PRE-COMPUTE AND CACHE EMBEDDINGS (MODIFIED for ESMC)
# ==============================================================================
print("\n" + "="*80)
print("... 2.5. PRE-COMPUTING AND CACHING EMBEDDINGS ...")

def generate_and_cache_embeddings(df: pd.DataFrame, cache_path="esmc_embeddings.pkl", batch_size=16):
    """
    Generates embeddings for all unique protein sequences and saves them to a cache file.
    If the cache file exists, it loads the embeddings from there.
    """
    if os.path.exists(cache_path):
        print(f"📂 Loading pre-computed embeddings from '{cache_path}'...")
        with open(cache_path, 'rb') as f:
            embedding_dict, embedding_size = pickle.load(f)
        print(f"✅ Loaded {len(embedding_dict)} embeddings. Embedding size: {embedding_size}")
        return embedding_dict, embedding_size

    print("🛠️ No cache found. Generating new embeddings...")

    unique_seqs = pd.concat([df['Target'], df['proteina']]).unique()
    print(f"Found {len(unique_seqs)} unique protein sequences to process.")

    generator = ESMCEmbeddingGenerator()
    embedding_size = generator.embedding_size

    embedding_dict = {}

    for i in tqdm(range(0, len(unique_seqs), batch_size), desc="Generating Embeddings"):
        batch_seqs = unique_seqs[i:i+batch_size]
        batch_embeddings = generator.get_embeddings(list(batch_seqs))

        for seq, emb in zip(batch_seqs, batch_embeddings):
            embedding_dict[seq] = emb

    print(f"💾 Saving {len(embedding_dict)} embeddings to '{cache_path}'...")
    with open(cache_path, 'wb') as f:
        pickle.dump((embedding_dict, embedding_size), f)

    del generator
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    print("✅ Embeddings generated and cached successfully.")
    return embedding_dict, embedding_size

# ==============================================================================
# 3. LOAD YOUR DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 3. LOADING DATA FROM Data.csv ...")

protein1_col = 'Target'
protein2_col = 'proteina'
target_col = 'Y'
pdb_col_raw = 'PDB'
subgroup_col = 'Subgroup'
source_dataset_col = 'Source Data Set' # NEW

try:
    # IMPORTANT: Update this path to your actual file location
    df_raw = pd.read_csv(r'/teamspace/studios/this_studio/PPB_Affinity_Sequences_Final (version 1).csv')
    print(f"Successfully loaded Data.csv with {len(df_raw)} rows.")
    df = df_raw[[protein1_col, protein2_col, target_col, pdb_col_raw,subgroup_col,source_dataset_col]].copy()
    df.rename(columns={
        protein1_col: 'Target',
        protein2_col: 'proteina',
        target_col: 'Y',
        pdb_col_raw: 'PDB',
        subgroup_col : 'Subgroup',
        source_dataset_col: 'SourceDataSet' # NEW
    }, inplace=True)

    print("\n--- Initial Data Check ---")
    print(df.head())
    missing_values = df.isnull().sum()
    print("\nMissing values per column:\n", missing_values)

    if missing_values.sum() > 0:
        df.dropna(subset=['Target', 'proteina', 'Y', 'PDB'], inplace=True)
        print(f"Dataset now has {len(df)} rows after cleaning.")

    df['Target'] = df['Target'].astype(str)
    df['proteina'] = df['proteina'].astype(str)
    df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
    df['PDB'] = df['PDB'].astype(str)
    df['Subgroup'] = df['Subgroup'].astype(str)
    df['SourceDataSet'] = df['SourceDataSet'].astype(str) # NEW

    df.dropna(subset=['Y'], inplace=True)

    print(f"\n✅ Data is loaded and cleaned. Final dataset size: {len(df)} samples.")
    df.info()

except FileNotFoundError:
    print("❌ ERROR: 'Data.csv' not found. Please ensure the path is correct.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB','Subgroup'])
except KeyError as e:
    print(f"❌ ERROR: Column {e} not found in 'Data.csv'. Please check column names.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB','Subgroup'])

# ==============================================================================
# 4. EXPERIMENT CONFIGURATION
# ==============================================================================
print("\n" + "="*80)
print("... 4. CONFIGURING THE EXPERIMENT ...")

CV_STRATEGY = 'GroupKFold'
N_SPLITS = 5
EPOCHS = 30
PATIENCE = 15
BATCH_SIZE = 16
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 512
DROPOUT = 0.1

# Output directory for results
RESULTS_DIR = "cv_results_frozen_esmc"
os.makedirs(RESULTS_DIR, exist_ok=True)
print(f"Results will be saved in: {RESULTS_DIR}")

print(f"✅ Configuration set. Using CV Strategy: {CV_STRATEGY}")
print(f"   Epochs: {EPOCHS}, Patience: {PATIENCE}")
print(f"   Training with a frozen ESMC backbone. Only the projection head will be trained.")

# ==============================================================================
# 5. ADAPTED MODEL AND DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 5. DEFINING ADAPTED MODEL AND DATASET ...")

class BALMProjectionHead(nn.Module):
    def __init__(self, embedding_size, projected_size, projected_dropout):
        super().__init__()
        self.protein_projection = nn.Linear(embedding_size, projected_size)
        self.proteina_projection = nn.Linear(embedding_size, projected_size)
        self.dropout = nn.Dropout(projected_dropout)
        self.loss_fn = nn.MSELoss()
        print(f"✅ BALMProjectionHead initialized with embedding_size: {embedding_size}")

    def forward(self, protein_embedding, proteina_embedding, labels=None):
        protein_embedding = self.dropout(protein_embedding)
        proteina_embedding = self.dropout(proteina_embedding)
        protein_projected = self.protein_projection(protein_embedding)
        proteina_projected = self.proteina_projection(proteina_embedding)
        protein_projected = F.normalize(protein_projected, p=2, dim=1)
        proteina_projected = F.normalize(proteina_projected, p=2, dim=1)
        cosine_similarity = F.cosine_similarity(protein_projected, proteina_projected)
        cosine_similarity = torch.clamp(cosine_similarity, -0.9999, 0.9999)
        output = {"cosine_similarity": cosine_similarity}
        if labels is not None:
            loss = self.loss_fn(cosine_similarity, labels)
            output["loss"] = loss
        return output

class BALMForRegression(nn.Module):
    def __init__(self, embedding_size, projected_size, projected_dropout, pkd_bounds):
        super().__init__()
        self.projection_head = BALMProjectionHead(embedding_size, projected_size, projected_dropout)
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def forward(self, batch_input):
        protein_emb = batch_input["protein_embedding"]
        proteina_emb = batch_input["proteina_embedding"]
        labels = batch_input.get("labels")

        proj_output = self.projection_head(protein_emb, proteina_emb, labels)

        # Return a unified dictionary
        output = {
            "cosine_similarity": proj_output["cosine_similarity"],
            "original_pkds": batch_input["original_pkds"],
            "pdb_groups": batch_input["pdb_groups"],
            "subgroups": batch_input["subgroups"],
            "source_dataset": batch_input["source_dataset"]
        }
        if "loss" in proj_output:
            output["loss"] = proj_output["loss"]
        return output

class ProteinPairEmbeddingDataset(Dataset):
    def __init__(self, dataframe, embedding_dict, pkd_bounds):
        self.data = dataframe.reset_index(drop=True)
        self.embedding_dict = embedding_dict
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        original_pkd = float(row['Y'])
        scaled_label = ((original_pkd - self.pkd_lower) / self.pkd_range) * 2 - 1

        protein_seq = row["Target"]
        proteina_seq = row["proteina"]

        return {
            "protein_embedding": self.embedding_dict[protein_seq],
            "proteina_embedding": self.embedding_dict[proteina_seq],
            "labels": torch.tensor(scaled_label, dtype=torch.float32),
            "original_pkds": torch.tensor(original_pkd, dtype=torch.float32),
            "pdb_groups": row["PDB"],
            "subgroups": row["Subgroup"],
            "source_dataset": row["SourceDataSet"] # NEW
        }

def collate_fn_embeddings(batch):
    return {
        "protein_embedding": torch.stack([item['protein_embedding'] for item in batch]),
        "proteina_embedding": torch.stack([item['proteina_embedding'] for item in batch]),
        "labels": torch.stack([item['labels'] for item in batch]),
        "original_pkds": torch.stack([item['original_pkds'] for item in batch]),
        "pdb_groups": [item['pdb_groups'] for item in batch],
        "subgroups": [item['subgroups'] for item in batch],
        "source_dataset": [item['source_dataset'] for item in batch] # NEW
    }

# ==============================================================================
# 6. METRICS AND EVALUATION FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6. DEFINING METRICS AND EVALUATION ...")

def concordance_index(y_true, y_pred):
    """Calculates the Concordance Index (C-index) for regression."""
    try:
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        n = len(y_true)
        if n < 2: return 1.0
        concordant, discordant, tied_pred = 0, 0, 0
        for i in range(n):
            for j in range(i + 1, n):
                if y_true[i] == y_true[j]: continue
                if y_pred[i] == y_pred[j]:
                    tied_pred += 1
                    continue
                if (y_pred[i] > y_pred[j] and y_true[i] > y_true[j]) or \
                   (y_pred[i] < y_pred[j] and y_true[i] < y_true[j]):
                    concordant += 1
                else:
                    discordant += 1
        total_pairs = concordant + discordant + tied_pred
        return 1.0 if total_pairs == 0 else (concordant + 0.5 * tied_pred) / total_pairs
    except Exception:
        return np.nan

def evaluate_model(model, data_loader, device, pkd_bounds):
    model.eval()
    all_labels, all_preds_pkd, all_pdb_groups, all_subgroups, all_sources = [], [], [], [], []
    pkd_lower, pkd_range = pkd_bounds[0], pkd_bounds[1] - pkd_bounds[0]

    with torch.no_grad():
        for batch in data_loader:
            # Move Tensors to GPU, keep metadata lists on CPU
            batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}

            outputs = model(batch_gpu)
            cosine_sim = outputs['cosine_similarity'].cpu()
            # Inverse scaling
            preds_pkd = ((cosine_sim + 1) / 2) * pkd_range + pkd_lower

            all_labels.extend(batch['original_pkds'].numpy())
            all_preds_pkd.extend(preds_pkd.numpy())
            all_pdb_groups.extend(batch['pdb_groups'])
            all_subgroups.extend(batch['subgroups'])
            all_sources.extend(batch['source_dataset'])

    # Standardize names to match the return statement
    y_true, y_pred = np.array(all_labels), np.array(all_preds_pkd)

    mask = ~np.isnan(y_true) & ~np.isnan(y_pred)
    labels_masked, preds_masked = y_true[mask], y_pred[mask]

    if len(labels_masked) < 2:
        metrics = {"rmse": np.nan, "pearson": np.nan, "spearman": np.nan, "ci": np.nan}
    else:
        metrics = {
            'rmse': np.sqrt(mean_squared_error(labels_masked, preds_masked)),
            'pearson': pearsonr(labels_masked, preds_masked)[0],
            'spearman': spearmanr(labels_masked, preds_masked)[0],
            'ci': concordance_index(labels_masked, preds_masked)
        }

    # Return exactly 6 items
    return metrics, y_true, y_pred, np.array(all_pdb_groups), np.array(all_subgroups), np.array(all_sources)

def plot_regression(y_true, y_pred, metrics, title, filename=None):
    plt.figure(figsize=(8, 8))
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, edgecolor='k', s=50)
    min_val, max_val = min(np.min(y_true), np.min(y_pred)), max(np.max(y_true), np.max(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Identity Line (y=x)')
    metrics_text = f"RMSE: {metrics['rmse']:.4f}\nPearson: {metrics['pearson']:.4f}\nSpearman: {metrics['spearman']:.4f}\nCI: {metrics['ci']:.4f}"
    plt.text(0.05, 0.95, metrics_text, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))
    plt.title(title, fontsize=16)
    plt.xlabel("Actual pKd", fontsize=14)
    plt.ylabel("Predicted pKd", fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    if filename:
        plt.savefig(filename)
        print(f"Plot saved to {filename}")
    plt.show()

# ==============================================================================
# 7. CROSS-VALIDATION TRAINING (MODIFIED for Frozen ESMC)
# ==============================================================================
print("\n" + "="*80)
print("... 7. RUNNING 'COLD TARGET' (GROUPKFOLD) STRATEGY WITH FROZEN ESMC ...")

def get_groupkfold_splits(df: pd.DataFrame, n_splits: int):
    print(f"\n❄️ Preparing {n_splits}-Fold **GroupKFold (Cold Target)** splits...")
    gkf = GroupKFold(n_splits=n_splits)
    groups = df['PDB'].factorize()[0]
    for fold, (train_idx, test_idx) in enumerate(gkf.split(df, groups=groups)):
        train_targets = set(df.iloc[train_idx]['PDB'].unique())
        test_targets = set(df.iloc[test_idx]['PDB'].unique())
        if not train_targets.intersection(test_targets):
            print(f"  ✅ Fold {fold+1}: No 'PDB' group leakage detected.")
        else:
            print(f"  ❌ Fold {fold+1} FAILED: PDB group leakage detected!")
        yield train_idx, test_idx

def run_cold_target_cv_frozen_esmc():
    if df.empty:
        print("⚠️ Skipping training due to empty dataframe.")
        return

    # --- MODIFIED: Use the correct cache path for ESMC embeddings ---
    protein_embeddings, embedding_size = generate_and_cache_embeddings(df, cache_path="esmc_embeddings.pkl")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    PKD_LOWER_BOUND, PKD_UPPER_BOUND = df['Y'].min(), df['Y'].max()
    PKD_BOUNDS = (PKD_LOWER_BOUND, PKD_UPPER_BOUND)
    print(f"Global pKd bounds for scaling: [{PKD_LOWER_BOUND:.2f}, {PKD_UPPER_BOUND:.2f}]")

    print(f"\n🚀 Starting {N_SPLITS}-Fold **{CV_STRATEGY}** Cross-Validation with Frozen ESMC backbone...")

    splits = get_groupkfold_splits(df, N_SPLITS)
    fold_results, all_folds_y_true, all_folds_y_pred = [], [], []

    for fold, (train_ids, val_ids) in enumerate(splits):
        fold_num = fold + 1
        print(f"\n===== STRATEGY: {CV_STRATEGY} | FOLD {fold_num}/{N_SPLITS} =====")

        train_df, val_df = df.iloc[train_ids], df.iloc[val_ids]

        train_dataset = ProteinPairEmbeddingDataset(train_df, protein_embeddings, PKD_BOUNDS)
        val_dataset = ProteinPairEmbeddingDataset(val_df, protein_embeddings, PKD_BOUNDS)
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn_embeddings)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE*2, shuffle=False, collate_fn=collate_fn_embeddings)

        model = BALMForRegression(
            embedding_size=embedding_size,
            projected_size=PROJECTED_SIZE,
            projected_dropout=DROPOUT,
            pkd_bounds=PKD_BOUNDS
        ).to(device)

        optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)
        print(f"   Optimizing {sum(p.numel() for p in model.parameters() if p.requires_grad)} trainable parameters.")

        best_val_metric = float('inf')
        patience_counter = 0
        fold_model_path = os.path.join(RESULTS_DIR, f"best_model_fold_{fold_num}.pth")

        for epoch in range(EPOCHS):
            model.train()
            total_train_loss = 0.0
            progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} (Fold {fold_num})", leave=False)

            for batch in progress_bar:
                optimizer.zero_grad()
                batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
                batch_gpu['pdb_groups'] = batch['pdb_groups']
                batch_gpu['subgroups'] = batch['subgroups']

                outputs = model(batch_gpu)
                loss = outputs['loss']
                loss.backward()
                optimizer.step()
                total_train_loss += loss.item()
                progress_bar.set_postfix({'batch_loss': f"{loss.item():.4f}"})

            avg_train_loss = total_train_loss / len(train_loader)

            val_metrics, _, _, _, _, _ = evaluate_model(model, val_loader, device, PKD_BOUNDS)

            current_val_metric = val_metrics['rmse']

            print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {avg_train_loss:.4f} | Val RMSE: {current_val_metric:.4f} | Val Pearson: {val_metrics['pearson']:.4f}")

            if current_val_metric < best_val_metric:
                best_val_metric = current_val_metric
                patience_counter = 0
                torch.save(model.state_dict(), fold_model_path)
                print(f"   -> New best model for Fold {fold_num} saved with Val RMSE: {best_val_metric:.4f}")
            else:
                patience_counter += 1

            if patience_counter >= PATIENCE:
                print(f"⏳ Early stopping triggered at epoch {epoch+1} for Fold {fold_num}.")
                break

        print(f"Fold {fold_num} training complete.")

        if os.path.exists(fold_model_path):
            model.load_state_dict(torch.load(fold_model_path))

        final_val_metrics, y_true_fold, y_pred_fold, pdb_groups_fold, subgroups_fold, sources_fold = evaluate_model(model, val_loader, device, PKD_BOUNDS)

        print(f"Fold {fold_num} Final Validation Metrics:")
        for key, value in final_val_metrics.items(): print(f"  - {key.upper()}: {value:.4f}")

        fold_results.append(final_val_metrics)
        all_folds_y_true.extend(y_true_fold)
        all_folds_y_pred.extend(y_pred_fold)

        fold_output_df = pd.DataFrame({
            'label': y_true_fold,
            'prediction': y_pred_fold,
            'residual': y_true_fold - y_pred_fold,
            'abs_residual': np.abs(y_true_fold - y_pred_fold),
            'PDB': pdb_groups_fold,
            'Subgroup': subgroups_fold,
            'Source Data Set': sources_fold # Ensure this is included
        })
        fold_output_filename = os.path.join(RESULTS_DIR, f"fold_{fold_num}_predictions.csv")
        fold_output_df.to_csv(fold_output_filename, index=False)
        print(f"Predictions for Fold {fold_num} saved to {fold_output_filename}")

    print(f"\n📈 Generating combined regression plot for {CV_STRATEGY}")
    combined_y_true, combined_y_pred = np.array(all_folds_y_true), np.array(all_folds_y_pred)
    overall_metrics = {
        'rmse': np.sqrt(mean_squared_error(combined_y_true, combined_y_pred)),
        'pearson': pearsonr(combined_y_true, combined_y_pred)[0],
        'spearman': spearmanr(combined_y_true, combined_y_pred)[0],
        'ci': concordance_index(combined_y_true, combined_y_pred)
    }

    plot_regression(
        combined_y_true, combined_y_pred, overall_metrics,
        title=f"Overall Regression for {CV_STRATEGY} ({N_SPLITS}-Fold CV) with Frozen ESMC",
        filename=os.path.join(RESULTS_DIR, "overall_regression_plot.png")
    )

    print(f"\n✅ Final Results Summary for {CV_STRATEGY} with Frozen ESMC:")
    results_df = pd.DataFrame(fold_results)
    summary_df = pd.DataFrame({'Mean': results_df.mean(), 'Std Dev': results_df.std()})
    print(summary_df)

    summary_filename = os.path.join(RESULTS_DIR, "cv_summary_metrics.csv")
    summary_df.to_csv(summary_filename)
    print(f"Overall CV summary metrics saved to {summary_filename}")


# --- Call the CV function ---
run_cold_target_cv_frozen_esmc()

# ==============================================================================
# 8. FINAL NOTES AND RECOMMENDATIONS
# ==============================================================================
print("\n" + "="*80)
print("... 8. FINAL NOTES AND RECOMMENDATIONS ...")
print("""
🎉 CONGRATULATIONS! You've successfully implemented BALM using pre-computed embeddings
from a **Frozen ESMC Backbone**!

📊 KEY ADVANTAGES OF THIS APPROACH:
• **Extreme Speed:** Training is incredibly fast because the large ESMC model is not used in the training loop.
• **Memory Efficiency:** Requires very little GPU memory during training, allowing for much larger batch sizes.
• **Simplicity:** The model architecture and training pipeline are simpler compared to fine-tuning approaches.
• **State-of-the-Art Representations:** You are now using embeddings from the ESMC family of models, which are specialized for protein representation learning.

📉 POTENTIAL LIMITATIONS:
• **Static Features:** The ESMC embeddings are fixed and do not adapt to your specific dataset during training. This might limit the model's performance on tasks that are very different from ESMC's pre-training objectives.
• **Less Flexibility:** Unlike fine-tuning, this approach cannot learn new sequence-level patterns; it can only learn how to combine the existing fixed features.

🚀 PERFORMANCE OPTIMIZATION TIPS:
1.  **Batch Size:** Since this approach is memory-efficient, you can significantly increase the `BATCH_SIZE` to speed up training and potentially get more stable gradient updates.
2.  **Projection Head Architecture:** The bottleneck is now the small projection head. You could experiment with a deeper or wider head (e.g., adding more linear layers and non-linearities) to see if it improves performance.
3.  **Learning Rate:** The optimal learning rate for training just the head might be different from the one used for fine-tuning. Experiment with values from 1e-4 to 1e-3.

🎯 NEXT STEPS:
1.  **Compare Results:** Directly compare the performance metrics (RMSE, Pearson, etc.) from this run with your previous ESM-2 or LoRA fine-tuning runs. This will quantify the benefit of using ESMC embeddings for your specific task.
2.  **Experiment with Different ESMC Models:** You can easily swap the `model_name` in the `ESMCEmbeddingGenerator` to a larger ESMC model (e.g., 'esmc_600m') and regenerate the embeddings to see how the base model size impacts performance.
""")
print("\n✅ Frozen ESMC experiment complete! Happy protein modeling! 🧬🤖")
```

```python
# ==============================================================================
# 0. IMPORTS AND INITIAL SETUP
# ==============================================================================
print("\n" + "="*80)
print("... 0. SCRIPT INITIALIZED FOR FROZEN ESMC EXPERIMENT ...")
print("    Training with a frozen ESMC backbone and a trainable projection head.")

# ==============================================================================
# 1. IMPORTS AND INITIAL SETUP
# ==============================================================================
import os
import sys
import time
import json
import random
import pickle
from pathlib import Path
from datetime import datetime

# Standard data science libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import warnings

# Machine learning libraries
from sklearn.model_selection import KFold, GroupKFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr

# PyTorch
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW

# --- MODIFIED: Imports for ESMC ---
from esm.models.esmc import ESMC
from esm.sdk.api import ESMProtein, LogitsConfig

from typing import List

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
warnings.filterwarnings('ignore', category=UserWarning)

# --- MODIFIED: Title reflects the use of a frozen ESMC model ---
print("🧬 BALM Adaptation with a Frozen ESMC Backbone")
print("⚡ Fast Training with 5-Fold Cross-Validation and Pre-computed Embeddings")
print("="*80)

def setup_reproducibility(seed=42):
    """Setup complete reproducibility."""
    print(f"🔒 Setting up reproducibility with seed {seed}")
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)
    print(f"✅ Reproducibility setup complete")

setup_reproducibility(42)

# ==============================================================================
# 2. ESMC EMBEDDING GENERATOR (MODIFIED for ESMC)
# ==============================================================================

class ESMCEmbeddingGenerator:
    """
    A class to load a frozen ESMC model and generate embeddings for a list of sequences.
    This is used for pre-computation.
    """
    def __init__(
        self,
        model_name="esmc_300m",
        device="auto"
    ):
        self.model_name = model_name

        if device == "auto":
            self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        else:
            self.device = torch.device(device)

        print(f"🔧 Initializing ESMCEmbeddingGenerator on {self.device}")

        self.dtype = torch.float16 if self.device.type == 'cuda' else torch.float32
        print(f"📥 Loading ESMC model: {model_name} with dtype: {self.dtype}")

        self.model = ESMC.from_pretrained(model_name).to(self.device)
        self.model.eval()

        print("   ⚠️ All base model weights are frozen for embedding generation.")
        for param in self.model.parameters():
            param.requires_grad = False

        # Determine embedding size by processing a dummy sequence
        with torch.no_grad():
            dummy_protein = ESMProtein(sequence="A")
            dummy_tensor = self.model.encode(dummy_protein)
            dummy_output = self.model.logits(
                dummy_tensor, LogitsConfig(sequence=True, return_embeddings=True)
            )
            self.embedding_size = dummy_output.embeddings.shape[-1]

        print(f"   ✅ Model loaded. Embedding size from dummy run: {self.embedding_size}")

    @torch.no_grad()
    def get_embeddings(self, sequences: List[str]):
        all_embeddings = []
        for seq in sequences:
            # CRITICAL: Replace '|' with '<s><s>' for multi-chain support
            processed_seq = seq.replace('|', '<s><s>')

            protein = ESMProtein(sequence=processed_seq)
            protein_tensor = self.model.encode(protein)
            logits_output = self.model.logits(
                protein_tensor, LogitsConfig(sequence=True, return_embeddings=True)
            )
            # Mean pooling over the full sequence (including internal separator tokens)
            mean_embedding = logits_output.embeddings.mean(dim=1).squeeze(0)
            all_embeddings.append(mean_embedding)

        return torch.stack(all_embeddings).to(torch.float32).cpu()


# ==============================================================================
# 2.5. PRE-COMPUTE AND CACHE EMBEDDINGS (MODIFIED for ESMC)
# ==============================================================================
print("\n" + "="*80)
print("... 2.5. PRE-COMPUTING AND CACHING EMBEDDINGS ...")

def generate_and_cache_embeddings(df: pd.DataFrame, cache_path="esmc_embeddings.pkl", batch_size=16):
    """
    Generates embeddings for all unique protein sequences and saves them to a cache file.
    If the cache file exists, it loads the embeddings from there.
    """
    if os.path.exists(cache_path):
        print(f"📂 Loading pre-computed embeddings from '{cache_path}'...")
        with open(cache_path, 'rb') as f:
            embedding_dict, embedding_size = pickle.load(f)
        print(f"✅ Loaded {len(embedding_dict)} embeddings. Embedding size: {embedding_size}")
        return embedding_dict, embedding_size

    print("🛠️ No cache found. Generating new embeddings...")

    unique_seqs = pd.concat([df['Target'], df['proteina']]).unique()
    print(f"Found {len(unique_seqs)} unique protein sequences to process.")

    generator = ESMCEmbeddingGenerator()
    embedding_size = generator.embedding_size

    embedding_dict = {}

    for i in tqdm(range(0, len(unique_seqs), batch_size), desc="Generating Embeddings"):
        batch_seqs = unique_seqs[i:i+batch_size]
        batch_embeddings = generator.get_embeddings(list(batch_seqs))

        for seq, emb in zip(batch_seqs, batch_embeddings):
            embedding_dict[seq] = emb

    print(f"💾 Saving {len(embedding_dict)} embeddings to '{cache_path}'...")
    with open(cache_path, 'wb') as f:
        pickle.dump((embedding_dict, embedding_size), f)

    del generator
    if torch.cuda.is_available():
        torch.cuda.empty_cache()

    print("✅ Embeddings generated and cached successfully.")
    return embedding_dict, embedding_size

# ==============================================================================
# 3. LOAD YOUR DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 3. LOADING DATA FROM Data.csv ...")

protein1_col = 'Target'
protein2_col = 'proteina'
target_col = 'Y'
pdb_col_raw = 'PDB'
subgroup_col = 'Subgroup'
source_dataset_col = 'Source Data Set' # NEW

try:
    # IMPORTANT: Update this path to your actual file location
    df_raw = pd.read_csv(r'/teamspace/studios/this_studio/PPB_Affinity_Sequences_Final (version 1).csv')
    print(f"Successfully loaded Data.csv with {len(df_raw)} rows.")
    df = df_raw[[protein1_col, protein2_col, target_col, pdb_col_raw,subgroup_col,source_dataset_col]].copy()
    df.rename(columns={
        protein1_col: 'Target',
        protein2_col: 'proteina',
        target_col: 'Y',
        pdb_col_raw: 'PDB',
        subgroup_col : 'Subgroup',
        source_dataset_col: 'SourceDataSet' # NEW
    }, inplace=True)

    print("\n--- Initial Data Check ---")
    print(df.head())
    missing_values = df.isnull().sum()
    print("\nMissing values per column:\n", missing_values)

    if missing_values.sum() > 0:
        df.dropna(subset=['Target', 'proteina', 'Y', 'PDB'], inplace=True)
        print(f"Dataset now has {len(df)} rows after cleaning.")

    df['Target'] = df['Target'].astype(str)
    df['proteina'] = df['proteina'].astype(str)
    df['Y'] = pd.to_numeric(df['Y'], errors='coerce')
    df['PDB'] = df['PDB'].astype(str)
    df['Subgroup'] = df['Subgroup'].astype(str)
    df['SourceDataSet'] = df['SourceDataSet'].astype(str) # NEW

    df.dropna(subset=['Y'], inplace=True)

    print(f"\n✅ Data is loaded and cleaned. Final dataset size: {len(df)} samples.")
    df.info()

except FileNotFoundError:
    print("❌ ERROR: 'Data.csv' not found. Please ensure the path is correct.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB','Subgroup'])
except KeyError as e:
    print(f"❌ ERROR: Column {e} not found in 'Data.csv'. Please check column names.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB','Subgroup'])

# ==============================================================================
# 4. EXPERIMENT CONFIGURATION
# ==============================================================================
print("\n" + "="*80)
print("... 4. CONFIGURING THE EXPERIMENT ...")

CV_STRATEGY = 'GroupKFold'
N_SPLITS = 5
EPOCHS = 30
PATIENCE = 15
BATCH_SIZE = 16
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 1024
DROPOUT = 0.1

# Output directory for results
RESULTS_DIR = "cv_results_frozen_esmc"
os.makedirs(RESULTS_DIR, exist_ok=True)
print(f"Results will be saved in: {RESULTS_DIR}")

print(f"✅ Configuration set. Using CV Strategy: {CV_STRATEGY}")
print(f"   Epochs: {EPOCHS}, Patience: {PATIENCE}")
print(f"   Training with a frozen ESMC backbone. Only the projection head will be trained.")

# ==============================================================================
# 5. ADAPTED MODEL AND DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 5. DEFINING ADAPTED MODEL AND DATASET ...")

class BALMProjectionHead(nn.Module):
    def __init__(self, embedding_size, projected_size, projected_dropout):
        super().__init__()
        self.protein_projection = nn.Linear(embedding_size, projected_size)
        self.proteina_projection = nn.Linear(embedding_size, projected_size)
        self.dropout = nn.Dropout(projected_dropout)
        self.loss_fn = nn.MSELoss()
        print(f"✅ BALMProjectionHead initialized with embedding_size: {embedding_size}")

    def forward(self, protein_embedding, proteina_embedding, labels=None):
        protein_embedding = self.dropout(protein_embedding)
        proteina_embedding = self.dropout(proteina_embedding)
        protein_projected = self.protein_projection(protein_embedding)
        proteina_projected = self.proteina_projection(proteina_embedding)
        protein_projected = F.normalize(protein_projected, p=2, dim=1)
        proteina_projected = F.normalize(proteina_projected, p=2, dim=1)
        cosine_similarity = F.cosine_similarity(protein_projected, proteina_projected)
        cosine_similarity = torch.clamp(cosine_similarity, -0.9999, 0.9999)
        output = {"cosine_similarity": cosine_similarity}
        if labels is not None:
            loss = self.loss_fn(cosine_similarity, labels)
            output["loss"] = loss
        return output

class BALMForRegression(nn.Module):
    def __init__(self, embedding_size, projected_size, projected_dropout, pkd_bounds):
        super().__init__()
        self.projection_head = BALMProjectionHead(embedding_size, projected_size, projected_dropout)
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def forward(self, batch_input):
        protein_emb = batch_input["protein_embedding"]
        proteina_emb = batch_input["proteina_embedding"]
        labels = batch_input.get("labels")

        proj_output = self.projection_head(protein_emb, proteina_emb, labels)

        # Return a unified dictionary
        output = {
            "cosine_similarity": proj_output["cosine_similarity"],
            "original_pkds": batch_input["original_pkds"],
            "pdb_groups": batch_input["pdb_groups"],
            "subgroups": batch_input["subgroups"],
            "source_dataset": batch_input["source_dataset"]
        }
        if "loss" in proj_output:
            output["loss"] = proj_output["loss"]
        return output

class ProteinPairEmbeddingDataset(Dataset):
    def __init__(self, dataframe, embedding_dict, pkd_bounds):
        self.data = dataframe.reset_index(drop=True)
        self.embedding_dict = embedding_dict
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        original_pkd = float(row['Y'])
        scaled_label = ((original_pkd - self.pkd_lower) / self.pkd_range) * 2 - 1

        protein_seq = row["Target"]
        proteina_seq = row["proteina"]

        return {
            "protein_embedding": self.embedding_dict[protein_seq],
            "proteina_embedding": self.embedding_dict[proteina_seq],
            "labels": torch.tensor(scaled_label, dtype=torch.float32),
            "original_pkds": torch.tensor(original_pkd, dtype=torch.float32),
            "pdb_groups": row["PDB"],
            "subgroups": row["Subgroup"],
            "source_dataset": row["SourceDataSet"] # NEW
        }

def collate_fn_embeddings(batch):
    return {
        "protein_embedding": torch.stack([item['protein_embedding'] for item in batch]),
        "proteina_embedding": torch.stack([item['proteina_embedding'] for item in batch]),
        "labels": torch.stack([item['labels'] for item in batch]),
        "original_pkds": torch.stack([item['original_pkds'] for item in batch]),
        "pdb_groups": [item['pdb_groups'] for item in batch],
        "subgroups": [item['subgroups'] for item in batch],
        "source_dataset": [item['source_dataset'] for item in batch] # NEW
    }

# ==============================================================================
# 6. METRICS AND EVALUATION FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6. DEFINING METRICS AND EVALUATION ...")

def concordance_index(y_true, y_pred):
    """Calculates the Concordance Index (C-index) for regression."""
    try:
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        n = len(y_true)
        if n < 2: return 1.0
        concordant, discordant, tied_pred = 0, 0, 0
        for i in range(n):
            for j in range(i + 1, n):
                if y_true[i] == y_true[j]: continue
                if y_pred[i] == y_pred[j]:
                    tied_pred += 1
                    continue
                if (y_pred[i] > y_pred[j] and y_true[i] > y_true[j]) or \
                   (y_pred[i] < y_pred[j] and y_true[i] < y_true[j]):
                    concordant += 1
                else:
                    discordant += 1
        total_pairs = concordant + discordant + tied_pred
        return 1.0 if total_pairs == 0 else (concordant + 0.5 * tied_pred) / total_pairs
    except Exception:
        return np.nan

def evaluate_model(model, data_loader, device, pkd_bounds):
    model.eval()
    all_labels, all_preds_pkd, all_pdb_groups, all_subgroups, all_sources = [], [], [], [], []
    pkd_lower, pkd_range = pkd_bounds[0], pkd_bounds[1] - pkd_bounds[0]

    with torch.no_grad():
        for batch in data_loader:
            # Move Tensors to GPU, keep metadata lists on CPU
            batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}

            outputs = model(batch_gpu)
            cosine_sim = outputs['cosine_similarity'].cpu()
            # Inverse scaling
            preds_pkd = ((cosine_sim + 1) / 2) * pkd_range + pkd_lower

            all_labels.extend(batch['original_pkds'].numpy())
            all_preds_pkd.extend(preds_pkd.numpy())
            all_pdb_groups.extend(batch['pdb_groups'])
            all_subgroups.extend(batch['subgroups'])
            all_sources.extend(batch['source_dataset'])

    # Standardize names to match the return statement
    y_true, y_pred = np.array(all_labels), np.array(all_preds_pkd)

    mask = ~np.isnan(y_true) & ~np.isnan(y_pred)
    labels_masked, preds_masked = y_true[mask], y_pred[mask]

    if len(labels_masked) < 2:
        metrics = {"rmse": np.nan, "pearson": np.nan, "spearman": np.nan, "ci": np.nan}
    else:
        metrics = {
            'rmse': np.sqrt(mean_squared_error(labels_masked, preds_masked)),
            'pearson': pearsonr(labels_masked, preds_masked)[0],
            'spearman': spearmanr(labels_masked, preds_masked)[0],
            'ci': concordance_index(labels_masked, preds_masked)
        }

    # Return exactly 6 items
    return metrics, y_true, y_pred, np.array(all_pdb_groups), np.array(all_subgroups), np.array(all_sources)

def plot_regression(y_true, y_pred, metrics, title, filename=None):
    plt.figure(figsize=(8, 8))
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, edgecolor='k', s=50)
    min_val, max_val = min(np.min(y_true), np.min(y_pred)), max(np.max(y_true), np.max(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Identity Line (y=x)')
    metrics_text = f"RMSE: {metrics['rmse']:.4f}\nPearson: {metrics['pearson']:.4f}\nSpearman: {metrics['spearman']:.4f}\nCI: {metrics['ci']:.4f}"
    plt.text(0.05, 0.95, metrics_text, transform=plt.gca().transAxes, fontsize=12, verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))
    plt.title(title, fontsize=16)
    plt.xlabel("Actual pKd", fontsize=14)
    plt.ylabel("Predicted pKd", fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    if filename:
        plt.savefig(filename)
        print(f"Plot saved to {filename}")
    plt.show()

# ==============================================================================
# 7. CROSS-VALIDATION TRAINING (MODIFIED for Frozen ESMC)
# ==============================================================================
print("\n" + "="*80)
print("... 7. RUNNING 'COLD TARGET' (GROUPKFOLD) STRATEGY WITH FROZEN ESMC ...")

def get_groupkfold_splits(df: pd.DataFrame, n_splits: int):
    print(f"\n❄️ Preparing {n_splits}-Fold **GroupKFold (Cold Target)** splits...")
    gkf = GroupKFold(n_splits=n_splits)
    groups = df['PDB'].factorize()[0]
    for fold, (train_idx, test_idx) in enumerate(gkf.split(df, groups=groups)):
        train_targets = set(df.iloc[train_idx]['PDB'].unique())
        test_targets = set(df.iloc[test_idx]['PDB'].unique())
        if not train_targets.intersection(test_targets):
            print(f"  ✅ Fold {fold+1}: No 'PDB' group leakage detected.")
        else:
            print(f"  ❌ Fold {fold+1} FAILED: PDB group leakage detected!")
        yield train_idx, test_idx

def run_cold_target_cv_frozen_esmc():
    if df.empty:
        print("⚠️ Skipping training due to empty dataframe.")
        return

    # --- MODIFIED: Use the correct cache path for ESMC embeddings ---
    protein_embeddings, embedding_size = generate_and_cache_embeddings(df, cache_path="esmc_embeddings.pkl")

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    PKD_LOWER_BOUND, PKD_UPPER_BOUND = df['Y'].min(), df['Y'].max()
    PKD_BOUNDS = (PKD_LOWER_BOUND, PKD_UPPER_BOUND)
    print(f"Global pKd bounds for scaling: [{PKD_LOWER_BOUND:.2f}, {PKD_UPPER_BOUND:.2f}]")

    print(f"\n🚀 Starting {N_SPLITS}-Fold **{CV_STRATEGY}** Cross-Validation with Frozen ESMC backbone...")

    splits = get_groupkfold_splits(df, N_SPLITS)
    fold_results, all_folds_y_true, all_folds_y_pred = [], [], []

    for fold, (train_ids, val_ids) in enumerate(splits):
        fold_num = fold + 1
        print(f"\n===== STRATEGY: {CV_STRATEGY} | FOLD {fold_num}/{N_SPLITS} =====")

        train_df, val_df = df.iloc[train_ids], df.iloc[val_ids]

        train_dataset = ProteinPairEmbeddingDataset(train_df, protein_embeddings, PKD_BOUNDS)
        val_dataset = ProteinPairEmbeddingDataset(val_df, protein_embeddings, PKD_BOUNDS)
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn_embeddings)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE*2, shuffle=False, collate_fn=collate_fn_embeddings)

        model = BALMForRegression(
            embedding_size=embedding_size,
            projected_size=PROJECTED_SIZE,
            projected_dropout=DROPOUT,
            pkd_bounds=PKD_BOUNDS
        ).to(device)

        optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)
        print(f"   Optimizing {sum(p.numel() for p in model.parameters() if p.requires_grad)} trainable parameters.")

        best_val_metric = float('inf')
        patience_counter = 0
        fold_model_path = os.path.join(RESULTS_DIR, f"best_model_fold_{fold_num}.pth")

        for epoch in range(EPOCHS):
            model.train()
            total_train_loss = 0.0
            progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS} (Fold {fold_num})", leave=False)

            for batch in progress_bar:
                optimizer.zero_grad()
                batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
                batch_gpu['pdb_groups'] = batch['pdb_groups']
                batch_gpu['subgroups'] = batch['subgroups']

                outputs = model(batch_gpu)
                loss = outputs['loss']
                loss.backward()
                optimizer.step()
                total_train_loss += loss.item()
                progress_bar.set_postfix({'batch_loss': f"{loss.item():.4f}"})

            avg_train_loss = total_train_loss / len(train_loader)

            val_metrics, _, _, _, _, _ = evaluate_model(model, val_loader, device, PKD_BOUNDS)

            current_val_metric = val_metrics['rmse']

            print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {avg_train_loss:.4f} | Val RMSE: {current_val_metric:.4f} | Val Pearson: {val_metrics['pearson']:.4f}")

            if current_val_metric < best_val_metric:
                best_val_metric = current_val_metric
                patience_counter = 0
                torch.save(model.state_dict(), fold_model_path)
                print(f"   -> New best model for Fold {fold_num} saved with Val RMSE: {best_val_metric:.4f}")
            else:
                patience_counter += 1

            if patience_counter >= PATIENCE:
                print(f"⏳ Early stopping triggered at epoch {epoch+1} for Fold {fold_num}.")
                break

        print(f"Fold {fold_num} training complete.")

        if os.path.exists(fold_model_path):
            model.load_state_dict(torch.load(fold_model_path))

        final_val_metrics, y_true_fold, y_pred_fold, pdb_groups_fold, subgroups_fold, sources_fold = evaluate_model(model, val_loader, device, PKD_BOUNDS)

        print(f"Fold {fold_num} Final Validation Metrics:")
        for key, value in final_val_metrics.items(): print(f"  - {key.upper()}: {value:.4f}")

        fold_results.append(final_val_metrics)
        all_folds_y_true.extend(y_true_fold)
        all_folds_y_pred.extend(y_pred_fold)

        fold_output_df = pd.DataFrame({
            'label': y_true_fold,
            'prediction': y_pred_fold,
            'residual': y_true_fold - y_pred_fold,
            'abs_residual': np.abs(y_true_fold - y_pred_fold),
            'PDB': pdb_groups_fold,
            'Subgroup': subgroups_fold,
            'Source Data Set': sources_fold # Ensure this is included
        })
        fold_output_filename = os.path.join(RESULTS_DIR, f"fold_{fold_num}_predictions.csv")
        fold_output_df.to_csv(fold_output_filename, index=False)
        print(f"Predictions for Fold {fold_num} saved to {fold_output_filename}")

    print(f"\n📈 Generating combined regression plot for {CV_STRATEGY}")
    combined_y_true, combined_y_pred = np.array(all_folds_y_true), np.array(all_folds_y_pred)
    overall_metrics = {
        'rmse': np.sqrt(mean_squared_error(combined_y_true, combined_y_pred)),
        'pearson': pearsonr(combined_y_true, combined_y_pred)[0],
        'spearman': spearmanr(combined_y_true, combined_y_pred)[0],
        'ci': concordance_index(combined_y_true, combined_y_pred)
    }

    plot_regression(
        combined_y_true, combined_y_pred, overall_metrics,
        title=f"Overall Regression for {CV_STRATEGY} ({N_SPLITS}-Fold CV) with Frozen ESMC",
        filename=os.path.join(RESULTS_DIR, "overall_regression_plot.png")
    )

    print(f"\n✅ Final Results Summary for {CV_STRATEGY} with Frozen ESMC:")
    results_df = pd.DataFrame(fold_results)
    summary_df = pd.DataFrame({'Mean': results_df.mean(), 'Std Dev': results_df.std()})
    print(summary_df)

    summary_filename = os.path.join(RESULTS_DIR, "cv_summary_metrics.csv")
    summary_df.to_csv(summary_filename)
    print(f"Overall CV summary metrics saved to {summary_filename}")


# --- Call the CV function ---
run_cold_target_cv_frozen_esmc()

# ==============================================================================
# 8. FINAL NOTES AND RECOMMENDATIONS
# ==============================================================================
print("\n" + "="*80)
print("... 8. FINAL NOTES AND RECOMMENDATIONS ...")
print("""
🎉 CONGRATULATIONS! You've successfully implemented BALM using pre-computed embeddings
from a **Frozen ESMC Backbone**!

📊 KEY ADVANTAGES OF THIS APPROACH:
• **Extreme Speed:** Training is incredibly fast because the large ESMC model is not used in the training loop.
• **Memory Efficiency:** Requires very little GPU memory during training, allowing for much larger batch sizes.
• **Simplicity:** The model architecture and training pipeline are simpler compared to fine-tuning approaches.
• **State-of-the-Art Representations:** You are now using embeddings from the ESMC family of models, which are specialized for protein representation learning.

📉 POTENTIAL LIMITATIONS:
• **Static Features:** The ESMC embeddings are fixed and do not adapt to your specific dataset during training. This might limit the model's performance on tasks that are very different from ESMC's pre-training objectives.
• **Less Flexibility:** Unlike fine-tuning, this approach cannot learn new sequence-level patterns; it can only learn how to combine the existing fixed features.

🚀 PERFORMANCE OPTIMIZATION TIPS:
1.  **Batch Size:** Since this approach is memory-efficient, you can significantly increase the `BATCH_SIZE` to speed up training and potentially get more stable gradient updates.
2.  **Projection Head Architecture:** The bottleneck is now the small projection head. You could experiment with a deeper or wider head (e.g., adding more linear layers and non-linearities) to see if it improves performance.
3.  **Learning Rate:** The optimal learning rate for training just the head might be different from the one used for fine-tuning. Experiment with values from 1e-4 to 1e-3.

🎯 NEXT STEPS:
1.  **Compare Results:** Directly compare the performance metrics (RMSE, Pearson, etc.) from this run with your previous ESM-2 or LoRA fine-tuning runs. This will quantify the benefit of using ESMC embeddings for your specific task.
2.  **Experiment with Different ESMC Models:** You can easily swap the `model_name` in the `ESMCEmbeddingGenerator` to a larger ESMC model (e.g., 'esmc_600m') and regenerate the embeddings to see how the base model size impacts performance.
""")
print("\n✅ Frozen ESMC experiment complete! Happy protein modeling! 🧬🤖")
```
