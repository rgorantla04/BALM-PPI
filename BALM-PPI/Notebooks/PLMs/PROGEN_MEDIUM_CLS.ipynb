```python
# ==============================================================================
# 1. IMPORTS AND INITIAL SETUP
# ==============================================================================
import os
import sys
import time
import json
import random
import pickle
from pathlib import Path
from datetime import datetime
from copy import deepcopy

# Standard data science libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import warnings

# Machine learning libraries
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr

# PyTorch and Transformers
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import AutoModelForCausalLM, AutoTokenizer, get_linear_schedule_with_warmup

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
warnings.filterwarnings('ignore', category=UserWarning)

print("🧬 BALM Adaptation for ProGen-2 Cached Embeddings")
print("⚡ Fast Training with 5-Fold Cross-Validation")
print("="*80)

def setup_reproducibility(seed=42):
    """Setup complete reproducibility."""
    print(f"🔒 Setting up reproducibility with seed {seed}")
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)
    print(f"✅ Reproducibility setup complete")

setup_reproducibility(42)

# ==============================================================================
# 2. PROGEN-2 EMBEDDING EXTRACTOR (CORRECTED WITH CACHE METHODS)
# ==============================================================================

class FastEmbeddingExtractor:
    def __init__(self, model_name="hugohrban/progen2-medium", max_seq_len=1024, batch_size=16, device="auto"):
        self.model_name, self.max_seq_len, self.batch_size = model_name, max_seq_len, batch_size
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu") if device == "auto" else torch.device(device)
        self.dtype = torch.float16 if self.device.type == 'cuda' else torch.float32

        print(f"🔧 Initializing Extractor on {self.device}")
        
        # Standard way to load with dtype
        self.model = AutoModelForCausalLM.from_pretrained(
            model_name, 
            trust_remote_code=True, 
            torch_dtype=self.dtype 
        ).to(self.device).eval()
        
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token
        self.tokenizer.padding_side = 'left'

        self.embedding_size = self.model.config.n_embd
        self.embedding_cache = {}

    def compute_embeddings_batch(self, sequences):
        if not sequences: return torch.empty(0, self.embedding_size)

        # CRITICAL: Replace '|' with two EOS tokens for multi-chain support
        sep = self.tokenizer.eos_token
        processed_seqs = [s.replace('|', f"{sep}{sep}") for s in sequences]

        inputs = self.tokenizer(processed_seqs, padding=True, truncation=True, max_length=self.max_seq_len, return_tensors="pt")
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = self.model(**inputs, output_hidden_states=True)
            last_hidden_states = outputs.hidden_states[-1] # [Batch, Seq, Dim]
            
            # CRITICAL: Mean Pooling (required for concatenated sequences)
            mask = inputs['attention_mask'].unsqueeze(-1).expand(last_hidden_states.size()).float()
            sum_embeddings = torch.sum(last_hidden_states * mask, 1)
            sum_mask = torch.clamp(mask.sum(1), min=1e-9)
            embeddings = (sum_embeddings / sum_mask).cpu()

        return embeddings

    def get_embeddings(self, sequences, use_cache=True, show_progress=True):
        if isinstance(sequences, str): sequences = [sequences]
        embeddings, to_compute, to_compute_indices = [], [], []
        for i, seq in enumerate(sequences):
            if use_cache and seq in self.embedding_cache:
                embeddings.append(self.embedding_cache[seq])
            else:
                embeddings.append(None)
                to_compute.append(seq)
                to_compute_indices.append(i)
        if to_compute:
            if show_progress: print(f"🧮 Computing {len(to_compute)} new embeddings...")
            computed_embeddings = []
            pbar = tqdm(range(0, len(to_compute), self.batch_size), desc="Computing embeddings", disable=not show_progress)
            for start_idx in pbar:
                end_idx = min(start_idx + self.batch_size, len(to_compute))
                batch_sequences = to_compute[start_idx:end_idx]
                batch_embeddings = self.compute_embeddings_batch(batch_sequences)
                computed_embeddings.append(batch_embeddings)
                if use_cache:
                    for j, seq in enumerate(batch_sequences):
                        self.embedding_cache[seq] = batch_embeddings[j:j+1]
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
            if computed_embeddings:
                all_computed = torch.cat(computed_embeddings, dim=0)
                for i, orig_idx in enumerate(to_compute_indices):
                    embeddings[orig_idx] = all_computed[i:i+1]

        if not embeddings: return torch.empty(0, self.embedding_size)
        final_embeddings = torch.cat(embeddings, dim=0)
        if show_progress: print(f"✅ Generated embeddings: {final_embeddings.shape}")
        return final_embeddings

    # --- ADDED THESE MISSING METHODS ---
    def save_cache(self, cache_path):
        """Saves the current embedding cache to a file using pickle."""
        print(f"💾 Saving embedding cache to {cache_path}")
        with open(cache_path, 'wb') as f:
            pickle.dump(self.embedding_cache, f)
        print(f"✅ Saved {len(self.embedding_cache)} cached embeddings.")

    def load_cache(self, cache_path):
        """Loads an embedding cache from a file if it exists."""
        if os.path.exists(cache_path):
            print(f"📥 Loading embedding cache from {cache_path}")
            try:
                with open(cache_path, 'rb') as f:
                    self.embedding_cache = pickle.load(f)
                print(f"✅ Loaded {len(self.embedding_cache)} cached embeddings.")
                return True
            except (pickle.UnpicklingError, EOFError) as e:
                print(f"⚠️ Could not load cache file (it might be corrupted): {e}. A new cache will be created.")
                self.embedding_cache = {}
                return False
        return False
    # ------------------------------------

# ==============================================================================
# 3. LOAD YOUR DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 3. LOADING DATA FROM Data.csv ...")

protein1_col = 'Target'    
protein2_col = 'proteina'  
target_col = 'Y'         
pdb_col_raw = 'PDB'
source_dataset_col = 'Source Data Set' # NEW

try:
    df_raw = pd.read_csv(r'/teamspace/studios/this_studio/PPB_Affinity_Sequences_Final (version 1).csv')
    print(f"Successfully loaded Data.csv with {len(df_raw)} rows.")

    # Include Source Data Set in the selection
    df = df_raw[[protein1_col, protein2_col, target_col, pdb_col_raw, source_dataset_col]].copy()
    
    df.rename(columns={
        protein1_col: 'Target',
        protein2_col: 'proteina',
        target_col: 'Y',
        pdb_col_raw: 'PDB',
        source_dataset_col: 'SourceDataSet' # Rename for easier access
    }, inplace=True)

    # --- Data Validation ---
    missing_values = df.isnull().sum()
    if missing_values.sum() > 0:
        print(f"\nWarning: Found {missing_values.sum()} total missing values. Dropping rows with NaNs.")
        # FIX: Added 'subset=' keyword here
        df.dropna(subset=['Target', 'proteina', 'Y', 'PDB'], inplace=True)
        print(f"Dataset now has {len(df)} rows after cleaning.")

    # Ensure data types are correct
    df['Target'] = df['Target'].astype(str)
    df['proteina'] = df['proteina'].astype(str)
    df['SourceDataSet'] = df['SourceDataSet'].astype(str) # NEW
    df['Y'] = pd.to_numeric(df['Y'], errors='coerce') 
    df['PDB'] = df['PDB'].astype(str)
    df.dropna(subset=['Y'], inplace=True) 

    print(f"\n✅ Data is loaded and cleaned. Final dataset size: {len(df)} samples.")

except FileNotFoundError:
    print("❌ ERROR: 'Data.csv' not found.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB', 'SourceDataSet'])
except KeyError as e:
    print(f"❌ ERROR: Column {e} not found in 'Data.csv'.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB', 'SourceDataSet'])

# ==============================================================================
# 4. COMPUTE OR LOAD CACHED EMBEDDINGS
# ==============================================================================
print("\n" + "="*80)
print("... 4. MANAGING PROGEN-2 EMBEDDINGS CACHE ...")

# --- Define the path for your persistent embedding cache ---
CACHE_FILE = "progen2_embeddings_cache.pkl"

# Instantiate the extractor
# This is needed in both cases to get the model's embedding size and to hold the cache
extractor = FastEmbeddingExtractor(batch_size=32)

# Try to load the cache from the file
cache_loaded = extractor.load_cache(CACHE_FILE)

# Collect all unique sequences from the dataframe. This is needed to check if
# the loaded cache is sufficient or if we need to compute more embeddings.
unique_sequences = pd.concat([df['Target'], df['proteina']]).unique().tolist()
print(f"Found {len(unique_sequences)} unique protein sequences in the dataset.")

# Check if any sequences are missing from the loaded cache
if cache_loaded:
    cached_keys = set(extractor.embedding_cache.keys())
    missing_sequences = [seq for seq in unique_sequences if seq not in cached_keys]
    if not missing_sequences:
        print("✅ All required embeddings were successfully loaded from the cache file.")
    else:
        print(f"⚠️ Cache is missing {len(missing_sequences)} embeddings. They will be computed now.")
else:
    print(f"Cache file '{CACHE_FILE}' not found. All embeddings will be computed.")

# This single call will now use the loaded cache and only compute what's missing.
# If the cache was empty or didn't exist, it computes everything.
_ = extractor.get_embeddings(unique_sequences, show_progress=True)

# Save the updated cache back to the file if any new embeddings were computed.
# This ensures the cache file is always up-to-date with the current dataset.
if not cache_loaded or (locals().get('missing_sequences') and missing_sequences):
    extractor.save_cache(CACHE_FILE)

# The EMBEDDING_CACHE is now our master lookup table for the training process
EMBEDDING_CACHE = extractor.embedding_cache
embedding_size = extractor.embedding_size
print(f"✅ Embedding cache is ready with {len(EMBEDDING_CACHE)} entries.")
print(f"Embedding dimension: {embedding_size}")

# ==============================================================================
# 5. ADAPTED MODEL AND DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 5. DEFINING ADAPTED MODEL AND DATASET ...")

class BALMForEmbeddings(nn.Module):
    """
    Adapted BALM model that accepts pre-computed ProGen-2 embeddings.
    This version is self-contained and does not need base_model.py or utils.py.
    """
    def __init__(
        self,
        embedding_size=1024, # ProGen-2 Small
        projected_size=256,
        projected_dropout=0.1,
        relu_before_cosine=False
    ):
        super().__init__()
        self.embedding_size = embedding_size

        self.protein_projection = nn.Linear(embedding_size, projected_size)
        self.proteina_projection = nn.Linear(embedding_size, projected_size)
        self.dropout = nn.Dropout(projected_dropout)
        self.relu_before_cosine = relu_before_cosine
        self.loss_fn = nn.MSELoss()

        print("✅ BALMForEmbeddings model initialized.")

    def forward(self, batch_input):
        # This model expects embeddings directly, not sequences
        protein_embedding = batch_input["protein_embedding"]
        proteina_embedding = batch_input["proteina_embedding"]

        # Apply dropout for regularization
        protein_embedding = self.dropout(protein_embedding)
        proteina_embedding = self.dropout(proteina_embedding)

        # Project to shared latent space
        protein_projected = self.protein_projection(protein_embedding)
        proteina_projected = self.proteina_projection(proteina_embedding)

        if self.relu_before_cosine:
            protein_projected = F.relu(protein_projected)
            proteina_projected = F.relu(proteina_projected)

        # L2 normalize before cosine similarity for numerical stability
        protein_projected = F.normalize(protein_projected, p=2, dim=1)
        proteina_projected = F.normalize(proteina_projected, p=2, dim=1)

        # Compute cosine similarity
        cosine_similarity = F.cosine_similarity(protein_projected, proteina_projected)
        cosine_similarity = torch.clamp(cosine_similarity, -0.9999, 0.9999)

        forward_output = {"cosine_similarity": cosine_similarity}

        # Compute loss if labels are provided
        if "labels" in batch_input and batch_input["labels"] is not None:
            labels = batch_input["labels"]
            loss = self.loss_fn(cosine_similarity, labels)
            forward_output["loss"] = loss

        return forward_output

class ProteinPairEmbeddingDataset(Dataset):
    def __init__(self, dataframe, embedding_cache, pkd_bounds):
        self.data = dataframe.reset_index(drop=True)
        self.embedding_cache = embedding_cache
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        p1_emb = self.embedding_cache[row["Target"]].squeeze(0)
        p2_emb = self.embedding_cache[row["proteina"]].squeeze(0)
        scaled_label = ((float(row['Y']) - self.pkd_lower) / self.pkd_range) * 2 - 1

        return {
            "protein_embedding": p1_emb.float(),
            "proteina_embedding": p2_emb.float(),
            "labels": torch.tensor(scaled_label, dtype=torch.float32),
            "original_pkds": torch.tensor(row['Y'], dtype=torch.float32),
            "pdbs": row["PDB"],
            "sources": row["SourceDataSet"] # Matches Section 3 renaming
        }

def collate_fn_embeddings(batch):
    return {
        "protein_embedding": torch.stack([item['protein_embedding'] for item in batch]),
        "proteina_embedding": torch.stack([item['proteina_embedding'] for item in batch]),
        "labels": torch.stack([item['labels'] for item in batch]),
        "original_pkds": torch.stack([item['original_pkds'] for item in batch]),
        "pdbs": [item['pdbs'] for item in batch],
        "sources": [item['sources'] for item in batch]
    }


# ==============================================================================
# 6. METRICS AND EVALUATION FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6. DEFINING METRICS AND EVALUATION ...")

def concordance_index(y_true, y_pred):
    """Calculate concordance index."""
    try:
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        n = len(y_true)
        if n < 2: return 1.0 # Not enough pairs to compare

        concordant = 0
        discordant = 0
        tied_pred = 0

        for i in range(n):
            for j in range(i + 1, n):
                if y_true[i] == y_true[j]:
                    continue

                if y_pred[i] == y_pred[j]:
                    tied_pred += 1
                    continue

                if (y_pred[i] > y_pred[j] and y_true[i] > y_true[j]) or \
                   (y_pred[i] < y_pred[j] and y_true[i] < y_true[j]):
                    concordant += 1
                else:
                    discordant += 1

        if (concordant + discordant + tied_pred) == 0:
            return 1.0

        ci = (concordant + 0.5 * tied_pred) / (concordant + discordant + tied_pred)
        return ci
    except Exception:
        return np.nan

def evaluate_model(model, data_loader, device, pkd_bounds):
    """
    Evaluate the model.
    RETURNS: A tuple of (metrics_dict, actual_labels_array, predicted_labels_array, pdbs, sources)
    """
    model.eval()
    all_labels, all_preds, all_pdbs, all_srcs = [], [], [], []

    pkd_lower, pkd_upper = pkd_bounds
    pkd_range = pkd_upper - pkd_lower

    with torch.no_grad():
        for batch in data_loader:
            batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
            outputs = model(batch_gpu)
            cosine_sim = outputs['cosine_similarity'].cpu()
            
            # De-scale prediction back to pKd
            preds_pkd = ((cosine_sim + 1) / 2) * pkd_range + pkd_lower

            all_labels.extend(batch['original_pkds'].numpy())
            all_preds.extend(preds_pkd.numpy())
            all_pdbs.extend(batch['pdbs'])    
            all_srcs.extend(batch['sources']) 

    labels = np.array(all_labels)
    preds = np.array(all_preds)

    # Clean out any NaNs just in case
    mask = ~np.isnan(labels) & ~np.isnan(preds)
    clean_labels, clean_preds = labels[mask], preds[mask]

    if len(clean_labels) < 2:
        metrics = {"rmse": np.nan, "pearson": np.nan, "spearman": np.nan, "ci": np.nan}
    else:
        metrics = {
            'rmse': np.sqrt(mean_squared_error(clean_labels, clean_preds)),
            'pearson': pearsonr(clean_labels, clean_preds)[0],
            'spearman': spearmanr(clean_labels, clean_preds)[0],
            'ci': concordance_index(clean_labels, clean_preds)
        }

    return metrics, labels, preds, all_pdbs, all_srcs
# ==============================================================================
# 6.A: EXPERIMENT CONFIGURATION
# ==============================================================================
print("\n" + "="*80)
print("... 6.A: CONFIGURING THE EXPERIMENT ...")

# --- CHOOSE YOUR CROSS-VALIDATION STRATEGY ---
# Options:
# 'KFold'              : Standard random split. Fastest, but risks data leakage.
# 'GroupKFold'         : Cold-target split. Ensures a protein seen in the test set was never seen in the train set.
# 'SequenceSimilarity' : Cold-cluster split. Groups similar proteins and puts them in the same fold. Most rigorous, but VERY SLOW.

CV_STRATEGY = 'GroupKFold' # <-- CHANGE THIS TO 'KFold' or 'SequenceSimilarity'

# --- HYPERPARAMETERS ---
N_SPLITS = 5
EPOCHS = 30 # Increase for better performance, e.g., 50-100
BATCH_SIZE = 32
EARLY_STOPPING_PATIENCE = 15    # stop if no RMSE improvement for 10 epochs
EARLY_STOPPING_MIN_DELTA = 1e-4 # minimum improvement in RMSE to reset patience
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 256
DROPOUT = 0.1

print(f"✅ Configuration set. Using CV Strategy: {CV_STRATEGY}")

# ==============================================================================
# 6.B: ADVANCED CROSS-VALIDATION SPLITTING FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6.B: DEFINING ADVANCED CV SPLITTING LOGIC ...")

# Required imports for the new strategies
from sklearn.model_selection import GroupKFold
from Bio.Align import PairwiseAligner
from scipy.cluster.hierarchy import linkage, fcluster
from multiprocessing import Pool, cpu_count
from typing import List, Tuple

# --- Strategy 1: GroupKFold (Cold Target) ---

def get_groupkfold_splits(df: pd.DataFrame, n_splits: int):
    """
    Generates train/test splits using GroupKFold to ensure cold-target validation.
    Proteins in the 'Target' column are used as groups.
    """
    print(f"\n❄️ Preparing {n_splits}-Fold **GroupKFold (Cold Target)** splits...")
    grouping_column = 'PDB'
    groups = df[grouping_column].factorize()[0]
    gkf = GroupKFold(n_splits=n_splits)

    for fold, (train_idx, test_idx) in enumerate(gkf.split(df, groups=groups)):
        # Verification step
        train_targets = set(df.iloc[train_idx][grouping_column])
        test_targets = set(df.iloc[test_idx][grouping_column])
        overlap = train_targets.intersection(test_targets)

        if len(overlap) > 0:
            print(f"❌ Fold {fold+1} FAILED: Target leakage detected with {len(overlap)} overlapping proteins!")
            # In a strict setting, you might raise an error here. For now, we'll just warn.
        else:
            print(f"  ✅ Fold {fold+1}: No target leakage detected.")

        yield train_idx, test_idx

# --- Strategy 2: Sequence Similarity (Cold Cluster) ---

def compute_pairwise_similarity(args: Tuple[str, str, int, int]) -> Tuple[int, int, float]:
    """Helper for parallel similarity calculation."""
    seq1, seq2, i, j = args
    aligner = PairwiseAligner()
    aligner.mode = 'global'
    aligner.match_score = 1
    aligner.mismatch_score = -1
    aligner.open_gap_score = -1
    aligner.extend_gap_score = -0.5
    try:
        score = aligner.score(seq1, seq2)
        sim = score / max(len(seq1), len(seq2), 1)
    except Exception:
        sim = 0.0
    return i, j, sim

def get_sequence_similarity_splits(df: pd.DataFrame, n_splits: int, identity_threshold: float = 0.5):
    """
    Generates train/test splits based on sequence similarity clustering.
    This is computationally intensive.
    """
    print(f"\n🧬 Preparing {n_splits}-Fold **Sequence Similarity** splits (Threshold={identity_threshold})...")
    print("⚠️ This process can be very slow, especially the first time.")

    # 1. Extract unique sequences
    print("   Step 1/5: Extracting unique sequences...")
    unique_sequences = list(pd.concat([df['Target'], df['proteina']]).unique())
    seq_to_idx = {seq: i for i, seq in enumerate(unique_sequences)}
    n_seqs = len(unique_sequences)

    # 2. Compute similarity matrix
    print(f"   Step 2/5: Computing similarity matrix for {n_seqs} sequences (parallelized)...")
    similarity_matrix = np.identity(n_seqs)
    pairs = [(unique_sequences[i], unique_sequences[j], i, j) for i in range(n_seqs) for j in range(i + 1, n_seqs)]

    with Pool(processes=max(1, cpu_count() - 1)) as pool:
        results = list(tqdm(pool.imap(compute_pairwise_similarity, pairs), total=len(pairs), desc="Calculating Similarities"))

    for i, j, sim in results:
        similarity_matrix[i, j] = similarity_matrix[j, i] = sim

    # 3. Cluster sequences
    print("   Step 3/5: Clustering sequences...")
    distance_matrix = 1 - similarity_matrix
    # Ensure distances are non-negative
    distance_matrix[distance_matrix < 0] = 0
    condensed_distance_matrix = distance_matrix[np.triu_indices(n_seqs, k=1)]

    Z = linkage(condensed_distance_matrix, method='complete')
    clusters = fcluster(Z, t=(1 - identity_threshold), criterion='distance')
    seq_to_cluster_id = {seq: cluster_id for seq, cluster_id in zip(unique_sequences, clusters)}
    num_clusters = len(set(clusters))
    print(f"   Found {num_clusters} clusters.")

    # 4. Assign clusters to folds
    print("   Step 4/5: Assigning clusters to folds...")
    cluster_ids = np.unique(clusters)
    np.random.shuffle(cluster_ids) # Randomize cluster order before assigning

    # Assign each cluster ID to a fold
    cluster_to_fold = {cid: i % n_splits for i, cid in enumerate(cluster_ids)}

    # 5. Create final splits
    print("   Step 5/5: Generating final train/test indices...")
    fold_indices = [[] for _ in range(n_splits)]
    for i, row in tqdm(df.iterrows(), total=len(df), desc="Assigning Pairs to Folds"):
        target_cid = seq_to_cluster_id.get(row['Target'])
        proteina_cid = seq_to_cluster_id.get(row['proteina'])

        # A pair belongs to a fold if EITHER of its proteins' clusters belongs to that fold
        target_fold = cluster_to_fold.get(target_cid, -1)
        proteina_fold = cluster_to_fold.get(proteina_cid, -1)

        # If proteins are in different clusters assigned to different folds,
        # we can decide on a rule. Let's assign to the target's fold.
        # This is a simplification; more complex rules could be used.
        assigned_fold = target_fold if target_fold != -1 else proteina_fold

        if assigned_fold != -1:
            fold_indices[assigned_fold].append(i)

    splits = []
    all_indices = set(df.index)
    for i in range(n_splits):
        test_idx = np.array(fold_indices[i])
        train_idx = np.array(list(all_indices - set(test_idx)))
        print(f"   Fold {i+1} Split: {len(train_idx)} train, {len(test_idx)} test")
        splits.append((train_idx, test_idx))

    return splits

print("✅ Advanced CV splitting functions are ready.")

def plot_regression(y_true, y_pred, metrics, title):
    """
    Generates and displays a regression plot for a given fold.
    """
    plt.figure(figsize=(8, 8))

    # Scatter plot
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, edgecolor='k', s=50)

    # Identity line (y=x)
    max_val = max(np.max(y_true), np.max(y_pred))
    min_val = min(np.min(y_true), np.min(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Identity Line (y=x)')

    # Display metrics on the plot
    metrics_text = (
        f"RMSE: {metrics['rmse']:.4f}\n"
        f"Pearson: {metrics['pearson']:.4f}\n"
        f"Spearman: {metrics['spearman']:.4f}\n"
        f"CI: {metrics['ci']:.4f}"
    )
    plt.text(0.05, 0.95, metrics_text, transform=plt.gca().transAxes,
             fontsize=12, verticalalignment='top',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))

    plt.title(title, fontsize=16)
    plt.xlabel("Actual pKd", fontsize=14)
    plt.ylabel("Predicted pKd", fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

print("✅ Regression plotting function is ready.")

# ==============================================================================
# 7. COMPREHENSIVE CROSS-VALIDATION & BENCHMARKING (WITH AGGREGATED PLOTTING)
# ==============================================================================
print("\n" + "="*80)
print("... 7. STARTING COMPREHENSIVE CV BENCHMARKING ...")

# --- HYPERPARAMETERS & CONFIG ---
strategies_to_run = ['GroupKFold']
# WARNING: 'SequenceSimilarity' can be very slow.
# For a quick test, use: strategies_to_run = ['KFold', 'GroupKFold']

N_SPLITS = 5
EPOCHS = 30
BATCH_SIZE = 32
EARLY_STOPPING_PATIENCE = 15    # stop if no RMSE improvement for 10 epochs
EARLY_STOPPING_MIN_DELTA = 1e-4 # minimum improvement in RMSE to reset patience
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 256
DROPOUT = 0.1

# --- MASTER RESULTS DICTIONARY ---
all_strategy_results = {}

# --- OUTER LOOP: Iterate over each CV strategy ---
for strategy in strategies_to_run:

    # --- Setup based on chosen CV Strategy ---
    if strategy == 'GroupKFold':
        print(f"\n🚀 Starting {N_SPLITS}-Fold **GroupKFold** Cross-Validation ...")
        splits = get_groupkfold_splits(df, N_SPLITS)
    elif strategy == 'SequenceSimilarity':
        print(f"\n🚀 Starting {N_SPLITS}-Fold **Sequence Similarity** Cross-Validation ...")
        splits = get_sequence_similarity_splits(df, N_SPLITS)
    else: # Default to KFold
        print(f"\n🚀 Starting {N_SPLITS}-Fold **Standard KFold** Cross-Validation ...")
        kfold = KFold(n_splits=N_SPLITS, shuffle=True, random_state=42)
        splits = kfold.split(df)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # --- Initialize lists to store results for the current strategy ---
    fold_results = [] # For statistical summary (mean, std)
    all_folds_y_true = [] # To collect all true labels for the aggregate plot
    all_folds_y_pred = [] # To collect all predictions for the aggregate plot

    PKD_LOWER_BOUND = df['Y'].min()
    PKD_UPPER_BOUND = df['Y'].max()
    PKD_BOUNDS = (PKD_LOWER_BOUND, PKD_UPPER_BOUND)
    print(f"Global pKd bounds for scaling: [{PKD_LOWER_BOUND:.2f}, {PKD_UPPER_BOUND:.2f}]")

    # --- INNER LOOP: Iterate over the folds for the current strategy ---
    for fold, (train_ids, val_ids) in enumerate(splits):
        fold_num = fold + 1
        print(f"\n===== STRATEGY: {strategy} | FOLD {fold_num}/{N_SPLITS} =====")

        train_df, val_df = df.iloc[train_ids], df.iloc[val_ids]
        train_dataset = ProteinPairEmbeddingDataset(train_df, EMBEDDING_CACHE, PKD_BOUNDS)
        val_dataset = ProteinPairEmbeddingDataset(val_df, EMBEDDING_CACHE, PKD_BOUNDS)
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn_embeddings)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE*2, shuffle=False, collate_fn=collate_fn_embeddings)

        model = BALMForEmbeddings(
            embedding_size=embedding_size,
            projected_size=PROJECTED_SIZE,
            projected_dropout=DROPOUT
        ).to(device)
        optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)

        # Training loop with Early Stopping (monitors validation RMSE)
        best_rmse = float("inf")
        best_state = None
        patience_counter = 0

        for epoch in range(EPOCHS):
            model.train()
            progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}", leave=False)
            for batch in progress_bar:
                optimizer.zero_grad()
                batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
                outputs = model(batch_gpu)
                loss = outputs['loss']
                loss.backward()
                optimizer.step()
                progress_bar.set_postfix({'train_loss': f"{loss.item():.4f}"})

            # ---- evaluate on validation at end of each epoch ----
            val_metrics, _, _, _, _ = evaluate_model(model, val_loader, device, PKD_BOUNDS)

            current_rmse = val_metrics['rmse'] if val_metrics['rmse'] is not None else float("inf")

            print(f"Epoch {epoch+1}/{EPOCHS} | Val RMSE: {current_rmse:.4f} | Best RMSE: {best_rmse:.4f} | Patience: {patience_counter}/{EARLY_STOPPING_PATIENCE}")

            # ---- early stopping check (lower RMSE is better) ----
            if (best_rmse - current_rmse) > EARLY_STOPPING_MIN_DELTA:
                best_rmse = current_rmse
                best_state = deepcopy(model.state_dict())
                patience_counter = 0
            else:
                patience_counter += 1
                if patience_counter >= EARLY_STOPPING_PATIENCE:
                    print(f"⏹️ Early stopping at epoch {epoch+1}: no RMSE improvement > {EARLY_STOPPING_MIN_DELTA} for {EARLY_STOPPING_PATIENCE} epochs.")
                    break

        # restore best weights before final fold evaluation
        if best_state is not None:
            model.load_state_dict(best_state)

        val_metrics, yt, yp, fold_pdbs, fold_sources = evaluate_model(model, val_loader, device, PKD_BOUNDS)
        
        pd.DataFrame({
            'label': yt, 
            'prediction': yp, 
            'residual': yt - yp,
            'PDB': fold_pdbs, 
            'Source Data Set': fold_sources
        }).to_csv(f"fold_{fold_num}_predictions.csv", index=False)

        print(f"Fold {fold_num} training complete (best Val RMSE: {best_rmse:.4f}).")


        # Evaluate and collect results
        # ---- Evaluate on validation set for the fold and collect results ----
        val_metrics, y_true, y_pred, fold_pdbs, fold_sources = evaluate_model(model, val_loader, device, PKD_BOUNDS)
        # Ensure arrays are 1D numpy arrays (safety)
        y_true = np.array(y_true).ravel() if y_true is not None else np.array([])
        y_pred = np.array(y_pred).ravel() if y_pred is not None else np.array([])

        print(f"Fold {fold_num} Validation Metrics:")
        for key, value in val_metrics.items():
            # Guard printing if metric is NaN
            print(f"  - {key.upper()}: {value:.4f}" if (value == value) else f"  - {key.upper()}: NaN")

        # Debug: how many validation examples this fold produced
        print(f"  ➤ Fold {fold_num} validation samples: {len(y_true)}")

        # Append fold metrics even if predictions empty (so we keep fold-level tracking),
        # but append predictions only if non-empty to avoid later aggregation errors.
        fold_results.append(val_metrics)
        if len(y_true) > 0 and len(y_pred) > 0:
            all_folds_y_true.extend(y_true.tolist())
            all_folds_y_pred.extend(y_pred.tolist())
        else:
            print(f"  ⚠️ No validation predictions to aggregate for fold {fold_num}. Skipping aggregation for this fold.")


    # --- AFTER ALL FOLDS for the strategy are done, create the AGGREGATE PLOT ---
    print(f"\n📈 Generating combined regression plot for strategy: {strategy}")

    # Convert lists to numpy arrays
    combined_y_true = np.array(all_folds_y_true)
    combined_y_pred = np.array(all_folds_y_pred)

    # Calculate overall metrics on the combined out-of-sample predictions
    overall_metrics = {
        'rmse': np.sqrt(mean_squared_error(combined_y_true, combined_y_pred)),
        'pearson': pearsonr(combined_y_true, combined_y_pred)[0],
        'spearman': spearmanr(combined_y_true, combined_y_pred)[0],
        'ci': concordance_index(combined_y_true, combined_y_pred)
    }

    # Generate the single plot for the entire strategy
    plot_regression(
        combined_y_true,
        combined_y_pred,
        overall_metrics,
        title=f"Overall Regression for {strategy} ({N_SPLITS}-Fold CV)"
    )

    # Store the per-fold results for the final statistical comparison plot
    all_strategy_results[strategy] = pd.DataFrame(fold_results)
    print(f"\n✅ Completed all folds for strategy: {strategy}")


# ==============================================================================
# 8. FINAL RESULTS COMPARISON
# ==============================================================================
print("\n" + "="*80)
print("... 8. FINAL CROSS-VALIDATION RESULTS COMPARISON ...")

# --- First, print summary tables for each strategy ---
for strategy, results_df in all_strategy_results.items():
    print(f"\n--- Summary for Strategy: {strategy} ---")
    summary_df = pd.DataFrame({
        'Mean': results_df.mean(),
        'Std Dev': results_df.std()
    })
    print(summary_df)
    print("-" * 40)

# --- Now, create the comparison error bar plots ---

def plot_comparison_results(all_results):
    """
    Generates bar plots with error bars to compare CV strategies.
    """
    metrics_to_plot = ['rmse', 'pearson', 'spearman', 'ci']

    for metric in metrics_to_plot:
        plt.figure(figsize=(10, 6))

        strategy_names = list(all_results.keys())
        mean_values = [results_df[metric].mean() for results_df in all_results.values()]
        std_devs = [results_df[metric].std() for results_df in all_results.values()]

        bars = plt.bar(strategy_names, mean_values, yerr=std_devs, capsize=5,
                       color=sns.color_palette("viridis", len(strategy_names)), alpha=0.8)

        plt.ylabel(metric.upper(), fontsize=14)
        plt.title(f"Comparison of CV Strategies for {metric.upper()}", fontsize=16)
        plt.xticks(rotation=0, ha="center", fontsize=12)

        # Add values on top of bars
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.3f}', va='bottom' if yval >= 0 else 'top', ha='center')

        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

if not all_strategy_results:
    print("⚠️ No results were generated. Skipping final comparison plot.")
else:
    plot_comparison_results(all_strategy_results)

print("\n✅ Benchmark analysis complete.")
```

```python
# ==============================================================================
# 1. IMPORTS AND INITIAL SETUP
# ==============================================================================
import os
import sys
import time
import json
import random
import pickle
from pathlib import Path
from datetime import datetime
from copy import deepcopy

# Standard data science libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import warnings

# Machine learning libraries
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr

# PyTorch and Transformers
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import AutoModelForCausalLM, AutoTokenizer, get_linear_schedule_with_warmup

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
warnings.filterwarnings('ignore', category=UserWarning)

print("🧬 BALM Adaptation for ProGen-2 Cached Embeddings")
print("⚡ Fast Training with 5-Fold Cross-Validation")
print("="*80)

def setup_reproducibility(seed=42):
    """Setup complete reproducibility."""
    print(f"🔒 Setting up reproducibility with seed {seed}")
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)
    print(f"✅ Reproducibility setup complete")

setup_reproducibility(42)

# ==============================================================================
# 2. PROGEN-2 EMBEDDING EXTRACTOR (CORRECTED WITH CACHE METHODS)
# ==============================================================================

class FastEmbeddingExtractor:
    def __init__(self, model_name="hugohrban/progen2-medium", max_seq_len=1024, batch_size=16, device="auto"):
        self.model_name, self.max_seq_len, self.batch_size = model_name, max_seq_len, batch_size
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu") if device == "auto" else torch.device(device)
        self.dtype = torch.float16 if self.device.type == 'cuda' else torch.float32

        print(f"🔧 Initializing Extractor on {self.device}")
        
        # Standard way to load with dtype
        self.model = AutoModelForCausalLM.from_pretrained(
            model_name, 
            trust_remote_code=True, 
            torch_dtype=self.dtype 
        ).to(self.device).eval()
        
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token
        self.tokenizer.padding_side = 'left'

        self.embedding_size = self.model.config.n_embd
        self.embedding_cache = {}

    def compute_embeddings_batch(self, sequences):
        if not sequences: return torch.empty(0, self.embedding_size)

        # CRITICAL: Replace '|' with two EOS tokens for multi-chain support
        sep = self.tokenizer.eos_token
        processed_seqs = [s.replace('|', f"{sep}{sep}") for s in sequences]

        inputs = self.tokenizer(processed_seqs, padding=True, truncation=True, max_length=self.max_seq_len, return_tensors="pt")
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = self.model(**inputs, output_hidden_states=True)
            last_hidden_states = outputs.hidden_states[-1] # [Batch, Seq, Dim]
            
            # CRITICAL: Mean Pooling (required for concatenated sequences)
            mask = inputs['attention_mask'].unsqueeze(-1).expand(last_hidden_states.size()).float()
            sum_embeddings = torch.sum(last_hidden_states * mask, 1)
            sum_mask = torch.clamp(mask.sum(1), min=1e-9)
            embeddings = (sum_embeddings / sum_mask).cpu()

        return embeddings

    def get_embeddings(self, sequences, use_cache=True, show_progress=True):
        if isinstance(sequences, str): sequences = [sequences]
        embeddings, to_compute, to_compute_indices = [], [], []
        for i, seq in enumerate(sequences):
            if use_cache and seq in self.embedding_cache:
                embeddings.append(self.embedding_cache[seq])
            else:
                embeddings.append(None)
                to_compute.append(seq)
                to_compute_indices.append(i)
        if to_compute:
            if show_progress: print(f"🧮 Computing {len(to_compute)} new embeddings...")
            computed_embeddings = []
            pbar = tqdm(range(0, len(to_compute), self.batch_size), desc="Computing embeddings", disable=not show_progress)
            for start_idx in pbar:
                end_idx = min(start_idx + self.batch_size, len(to_compute))
                batch_sequences = to_compute[start_idx:end_idx]
                batch_embeddings = self.compute_embeddings_batch(batch_sequences)
                computed_embeddings.append(batch_embeddings)
                if use_cache:
                    for j, seq in enumerate(batch_sequences):
                        self.embedding_cache[seq] = batch_embeddings[j:j+1]
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
            if computed_embeddings:
                all_computed = torch.cat(computed_embeddings, dim=0)
                for i, orig_idx in enumerate(to_compute_indices):
                    embeddings[orig_idx] = all_computed[i:i+1]

        if not embeddings: return torch.empty(0, self.embedding_size)
        final_embeddings = torch.cat(embeddings, dim=0)
        if show_progress: print(f"✅ Generated embeddings: {final_embeddings.shape}")
        return final_embeddings

    # --- ADDED THESE MISSING METHODS ---
    def save_cache(self, cache_path):
        """Saves the current embedding cache to a file using pickle."""
        print(f"💾 Saving embedding cache to {cache_path}")
        with open(cache_path, 'wb') as f:
            pickle.dump(self.embedding_cache, f)
        print(f"✅ Saved {len(self.embedding_cache)} cached embeddings.")

    def load_cache(self, cache_path):
        """Loads an embedding cache from a file if it exists."""
        if os.path.exists(cache_path):
            print(f"📥 Loading embedding cache from {cache_path}")
            try:
                with open(cache_path, 'rb') as f:
                    self.embedding_cache = pickle.load(f)
                print(f"✅ Loaded {len(self.embedding_cache)} cached embeddings.")
                return True
            except (pickle.UnpicklingError, EOFError) as e:
                print(f"⚠️ Could not load cache file (it might be corrupted): {e}. A new cache will be created.")
                self.embedding_cache = {}
                return False
        return False
    # ------------------------------------

# ==============================================================================
# 3. LOAD YOUR DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 3. LOADING DATA FROM Data.csv ...")

protein1_col = 'Target'    
protein2_col = 'proteina'  
target_col = 'Y'         
pdb_col_raw = 'PDB'
source_dataset_col = 'Source Data Set' # NEW

try:
    df_raw = pd.read_csv(r'/teamspace/studios/this_studio/PPB_Affinity_Sequences_Final (version 1).csv')
    print(f"Successfully loaded Data.csv with {len(df_raw)} rows.")

    # Include Source Data Set in the selection
    df = df_raw[[protein1_col, protein2_col, target_col, pdb_col_raw, source_dataset_col]].copy()
    
    df.rename(columns={
        protein1_col: 'Target',
        protein2_col: 'proteina',
        target_col: 'Y',
        pdb_col_raw: 'PDB',
        source_dataset_col: 'SourceDataSet' # Rename for easier access
    }, inplace=True)

    # --- Data Validation ---
    missing_values = df.isnull().sum()
    if missing_values.sum() > 0:
        print(f"\nWarning: Found {missing_values.sum()} total missing values. Dropping rows with NaNs.")
        # FIX: Added 'subset=' keyword here
        df.dropna(subset=['Target', 'proteina', 'Y', 'PDB'], inplace=True)
        print(f"Dataset now has {len(df)} rows after cleaning.")

    # Ensure data types are correct
    df['Target'] = df['Target'].astype(str)
    df['proteina'] = df['proteina'].astype(str)
    df['SourceDataSet'] = df['SourceDataSet'].astype(str) # NEW
    df['Y'] = pd.to_numeric(df['Y'], errors='coerce') 
    df['PDB'] = df['PDB'].astype(str)
    df.dropna(subset=['Y'], inplace=True) 

    print(f"\n✅ Data is loaded and cleaned. Final dataset size: {len(df)} samples.")

except FileNotFoundError:
    print("❌ ERROR: 'Data.csv' not found.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB', 'SourceDataSet'])
except KeyError as e:
    print(f"❌ ERROR: Column {e} not found in 'Data.csv'.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB', 'SourceDataSet'])

# ==============================================================================
# 4. COMPUTE OR LOAD CACHED EMBEDDINGS
# ==============================================================================
print("\n" + "="*80)
print("... 4. MANAGING PROGEN-2 EMBEDDINGS CACHE ...")

# --- Define the path for your persistent embedding cache ---
CACHE_FILE = "progen2_embeddings_cache.pkl"

# Instantiate the extractor
# This is needed in both cases to get the model's embedding size and to hold the cache
extractor = FastEmbeddingExtractor(batch_size=32)

# Try to load the cache from the file
cache_loaded = extractor.load_cache(CACHE_FILE)

# Collect all unique sequences from the dataframe. This is needed to check if
# the loaded cache is sufficient or if we need to compute more embeddings.
unique_sequences = pd.concat([df['Target'], df['proteina']]).unique().tolist()
print(f"Found {len(unique_sequences)} unique protein sequences in the dataset.")

# Check if any sequences are missing from the loaded cache
if cache_loaded:
    cached_keys = set(extractor.embedding_cache.keys())
    missing_sequences = [seq for seq in unique_sequences if seq not in cached_keys]
    if not missing_sequences:
        print("✅ All required embeddings were successfully loaded from the cache file.")
    else:
        print(f"⚠️ Cache is missing {len(missing_sequences)} embeddings. They will be computed now.")
else:
    print(f"Cache file '{CACHE_FILE}' not found. All embeddings will be computed.")

# This single call will now use the loaded cache and only compute what's missing.
# If the cache was empty or didn't exist, it computes everything.
_ = extractor.get_embeddings(unique_sequences, show_progress=True)

# Save the updated cache back to the file if any new embeddings were computed.
# This ensures the cache file is always up-to-date with the current dataset.
if not cache_loaded or (locals().get('missing_sequences') and missing_sequences):
    extractor.save_cache(CACHE_FILE)

# The EMBEDDING_CACHE is now our master lookup table for the training process
EMBEDDING_CACHE = extractor.embedding_cache
embedding_size = extractor.embedding_size
print(f"✅ Embedding cache is ready with {len(EMBEDDING_CACHE)} entries.")
print(f"Embedding dimension: {embedding_size}")

# ==============================================================================
# 5. ADAPTED MODEL AND DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 5. DEFINING ADAPTED MODEL AND DATASET ...")

class BALMForEmbeddings(nn.Module):
    """
    Adapted BALM model that accepts pre-computed ProGen-2 embeddings.
    This version is self-contained and does not need base_model.py or utils.py.
    """
    def __init__(
        self,
        embedding_size=1024, # ProGen-2 Small
        projected_size=512,
        projected_dropout=0.1,
        relu_before_cosine=False
    ):
        super().__init__()
        self.embedding_size = embedding_size

        self.protein_projection = nn.Linear(embedding_size, projected_size)
        self.proteina_projection = nn.Linear(embedding_size, projected_size)
        self.dropout = nn.Dropout(projected_dropout)
        self.relu_before_cosine = relu_before_cosine
        self.loss_fn = nn.MSELoss()

        print("✅ BALMForEmbeddings model initialized.")

    def forward(self, batch_input):
        # This model expects embeddings directly, not sequences
        protein_embedding = batch_input["protein_embedding"]
        proteina_embedding = batch_input["proteina_embedding"]

        # Apply dropout for regularization
        protein_embedding = self.dropout(protein_embedding)
        proteina_embedding = self.dropout(proteina_embedding)

        # Project to shared latent space
        protein_projected = self.protein_projection(protein_embedding)
        proteina_projected = self.proteina_projection(proteina_embedding)

        if self.relu_before_cosine:
            protein_projected = F.relu(protein_projected)
            proteina_projected = F.relu(proteina_projected)

        # L2 normalize before cosine similarity for numerical stability
        protein_projected = F.normalize(protein_projected, p=2, dim=1)
        proteina_projected = F.normalize(proteina_projected, p=2, dim=1)

        # Compute cosine similarity
        cosine_similarity = F.cosine_similarity(protein_projected, proteina_projected)
        cosine_similarity = torch.clamp(cosine_similarity, -0.9999, 0.9999)

        forward_output = {"cosine_similarity": cosine_similarity}

        # Compute loss if labels are provided
        if "labels" in batch_input and batch_input["labels"] is not None:
            labels = batch_input["labels"]
            loss = self.loss_fn(cosine_similarity, labels)
            forward_output["loss"] = loss

        return forward_output

class ProteinPairEmbeddingDataset(Dataset):
    def __init__(self, dataframe, embedding_cache, pkd_bounds):
        self.data = dataframe.reset_index(drop=True)
        self.embedding_cache = embedding_cache
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        p1_emb = self.embedding_cache[row["Target"]].squeeze(0)
        p2_emb = self.embedding_cache[row["proteina"]].squeeze(0)
        scaled_label = ((float(row['Y']) - self.pkd_lower) / self.pkd_range) * 2 - 1

        return {
            "protein_embedding": p1_emb.float(),
            "proteina_embedding": p2_emb.float(),
            "labels": torch.tensor(scaled_label, dtype=torch.float32),
            "original_pkds": torch.tensor(row['Y'], dtype=torch.float32),
            "pdbs": row["PDB"],
            "sources": row["SourceDataSet"] # Matches Section 3 renaming
        }

def collate_fn_embeddings(batch):
    return {
        "protein_embedding": torch.stack([item['protein_embedding'] for item in batch]),
        "proteina_embedding": torch.stack([item['proteina_embedding'] for item in batch]),
        "labels": torch.stack([item['labels'] for item in batch]),
        "original_pkds": torch.stack([item['original_pkds'] for item in batch]),
        "pdbs": [item['pdbs'] for item in batch],
        "sources": [item['sources'] for item in batch]
    }


# ==============================================================================
# 6. METRICS AND EVALUATION FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6. DEFINING METRICS AND EVALUATION ...")

def concordance_index(y_true, y_pred):
    """Calculate concordance index."""
    try:
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        n = len(y_true)
        if n < 2: return 1.0 # Not enough pairs to compare

        concordant = 0
        discordant = 0
        tied_pred = 0

        for i in range(n):
            for j in range(i + 1, n):
                if y_true[i] == y_true[j]:
                    continue

                if y_pred[i] == y_pred[j]:
                    tied_pred += 1
                    continue

                if (y_pred[i] > y_pred[j] and y_true[i] > y_true[j]) or \
                   (y_pred[i] < y_pred[j] and y_true[i] < y_true[j]):
                    concordant += 1
                else:
                    discordant += 1

        if (concordant + discordant + tied_pred) == 0:
            return 1.0

        ci = (concordant + 0.5 * tied_pred) / (concordant + discordant + tied_pred)
        return ci
    except Exception:
        return np.nan

def evaluate_model(model, data_loader, device, pkd_bounds):
    """
    Evaluate the model.
    RETURNS: A tuple of (metrics_dict, actual_labels_array, predicted_labels_array, pdbs, sources)
    """
    model.eval()
    all_labels, all_preds, all_pdbs, all_srcs = [], [], [], []

    pkd_lower, pkd_upper = pkd_bounds
    pkd_range = pkd_upper - pkd_lower

    with torch.no_grad():
        for batch in data_loader:
            batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
            outputs = model(batch_gpu)
            cosine_sim = outputs['cosine_similarity'].cpu()
            
            # De-scale prediction back to pKd
            preds_pkd = ((cosine_sim + 1) / 2) * pkd_range + pkd_lower

            all_labels.extend(batch['original_pkds'].numpy())
            all_preds.extend(preds_pkd.numpy())
            all_pdbs.extend(batch['pdbs'])    
            all_srcs.extend(batch['sources']) 

    labels = np.array(all_labels)
    preds = np.array(all_preds)

    # Clean out any NaNs just in case
    mask = ~np.isnan(labels) & ~np.isnan(preds)
    clean_labels, clean_preds = labels[mask], preds[mask]

    if len(clean_labels) < 2:
        metrics = {"rmse": np.nan, "pearson": np.nan, "spearman": np.nan, "ci": np.nan}
    else:
        metrics = {
            'rmse': np.sqrt(mean_squared_error(clean_labels, clean_preds)),
            'pearson': pearsonr(clean_labels, clean_preds)[0],
            'spearman': spearmanr(clean_labels, clean_preds)[0],
            'ci': concordance_index(clean_labels, clean_preds)
        }

    return metrics, labels, preds, all_pdbs, all_srcs
# ==============================================================================
# 6.A: EXPERIMENT CONFIGURATION
# ==============================================================================
print("\n" + "="*80)
print("... 6.A: CONFIGURING THE EXPERIMENT ...")

# --- CHOOSE YOUR CROSS-VALIDATION STRATEGY ---
# Options:
# 'KFold'              : Standard random split. Fastest, but risks data leakage.
# 'GroupKFold'         : Cold-target split. Ensures a protein seen in the test set was never seen in the train set.
# 'SequenceSimilarity' : Cold-cluster split. Groups similar proteins and puts them in the same fold. Most rigorous, but VERY SLOW.

CV_STRATEGY = 'GroupKFold' # <-- CHANGE THIS TO 'KFold' or 'SequenceSimilarity'

# --- HYPERPARAMETERS ---
N_SPLITS = 5
EPOCHS = 30 # Increase for better performance, e.g., 50-100
BATCH_SIZE = 32
EARLY_STOPPING_PATIENCE = 15    # stop if no RMSE improvement for 10 epochs
EARLY_STOPPING_MIN_DELTA = 1e-4 # minimum improvement in RMSE to reset patience
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 512
DROPOUT = 0.1

print(f"✅ Configuration set. Using CV Strategy: {CV_STRATEGY}")

# ==============================================================================
# 6.B: ADVANCED CROSS-VALIDATION SPLITTING FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6.B: DEFINING ADVANCED CV SPLITTING LOGIC ...")

# Required imports for the new strategies
from sklearn.model_selection import GroupKFold
from Bio.Align import PairwiseAligner
from scipy.cluster.hierarchy import linkage, fcluster
from multiprocessing import Pool, cpu_count
from typing import List, Tuple

# --- Strategy 1: GroupKFold (Cold Target) ---

def get_groupkfold_splits(df: pd.DataFrame, n_splits: int):
    """
    Generates train/test splits using GroupKFold to ensure cold-target validation.
    Proteins in the 'Target' column are used as groups.
    """
    print(f"\n❄️ Preparing {n_splits}-Fold **GroupKFold (Cold Target)** splits...")
    grouping_column = 'PDB'
    groups = df[grouping_column].factorize()[0]
    gkf = GroupKFold(n_splits=n_splits)

    for fold, (train_idx, test_idx) in enumerate(gkf.split(df, groups=groups)):
        # Verification step
        train_targets = set(df.iloc[train_idx][grouping_column])
        test_targets = set(df.iloc[test_idx][grouping_column])
        overlap = train_targets.intersection(test_targets)

        if len(overlap) > 0:
            print(f"❌ Fold {fold+1} FAILED: Target leakage detected with {len(overlap)} overlapping proteins!")
            # In a strict setting, you might raise an error here. For now, we'll just warn.
        else:
            print(f"  ✅ Fold {fold+1}: No target leakage detected.")

        yield train_idx, test_idx

# --- Strategy 2: Sequence Similarity (Cold Cluster) ---

def compute_pairwise_similarity(args: Tuple[str, str, int, int]) -> Tuple[int, int, float]:
    """Helper for parallel similarity calculation."""
    seq1, seq2, i, j = args
    aligner = PairwiseAligner()
    aligner.mode = 'global'
    aligner.match_score = 1
    aligner.mismatch_score = -1
    aligner.open_gap_score = -1
    aligner.extend_gap_score = -0.5
    try:
        score = aligner.score(seq1, seq2)
        sim = score / max(len(seq1), len(seq2), 1)
    except Exception:
        sim = 0.0
    return i, j, sim

def get_sequence_similarity_splits(df: pd.DataFrame, n_splits: int, identity_threshold: float = 0.5):
    """
    Generates train/test splits based on sequence similarity clustering.
    This is computationally intensive.
    """
    print(f"\n🧬 Preparing {n_splits}-Fold **Sequence Similarity** splits (Threshold={identity_threshold})...")
    print("⚠️ This process can be very slow, especially the first time.")

    # 1. Extract unique sequences
    print("   Step 1/5: Extracting unique sequences...")
    unique_sequences = list(pd.concat([df['Target'], df['proteina']]).unique())
    seq_to_idx = {seq: i for i, seq in enumerate(unique_sequences)}
    n_seqs = len(unique_sequences)

    # 2. Compute similarity matrix
    print(f"   Step 2/5: Computing similarity matrix for {n_seqs} sequences (parallelized)...")
    similarity_matrix = np.identity(n_seqs)
    pairs = [(unique_sequences[i], unique_sequences[j], i, j) for i in range(n_seqs) for j in range(i + 1, n_seqs)]

    with Pool(processes=max(1, cpu_count() - 1)) as pool:
        results = list(tqdm(pool.imap(compute_pairwise_similarity, pairs), total=len(pairs), desc="Calculating Similarities"))

    for i, j, sim in results:
        similarity_matrix[i, j] = similarity_matrix[j, i] = sim

    # 3. Cluster sequences
    print("   Step 3/5: Clustering sequences...")
    distance_matrix = 1 - similarity_matrix
    # Ensure distances are non-negative
    distance_matrix[distance_matrix < 0] = 0
    condensed_distance_matrix = distance_matrix[np.triu_indices(n_seqs, k=1)]

    Z = linkage(condensed_distance_matrix, method='complete')
    clusters = fcluster(Z, t=(1 - identity_threshold), criterion='distance')
    seq_to_cluster_id = {seq: cluster_id for seq, cluster_id in zip(unique_sequences, clusters)}
    num_clusters = len(set(clusters))
    print(f"   Found {num_clusters} clusters.")

    # 4. Assign clusters to folds
    print("   Step 4/5: Assigning clusters to folds...")
    cluster_ids = np.unique(clusters)
    np.random.shuffle(cluster_ids) # Randomize cluster order before assigning

    # Assign each cluster ID to a fold
    cluster_to_fold = {cid: i % n_splits for i, cid in enumerate(cluster_ids)}

    # 5. Create final splits
    print("   Step 5/5: Generating final train/test indices...")
    fold_indices = [[] for _ in range(n_splits)]
    for i, row in tqdm(df.iterrows(), total=len(df), desc="Assigning Pairs to Folds"):
        target_cid = seq_to_cluster_id.get(row['Target'])
        proteina_cid = seq_to_cluster_id.get(row['proteina'])

        # A pair belongs to a fold if EITHER of its proteins' clusters belongs to that fold
        target_fold = cluster_to_fold.get(target_cid, -1)
        proteina_fold = cluster_to_fold.get(proteina_cid, -1)

        # If proteins are in different clusters assigned to different folds,
        # we can decide on a rule. Let's assign to the target's fold.
        # This is a simplification; more complex rules could be used.
        assigned_fold = target_fold if target_fold != -1 else proteina_fold

        if assigned_fold != -1:
            fold_indices[assigned_fold].append(i)

    splits = []
    all_indices = set(df.index)
    for i in range(n_splits):
        test_idx = np.array(fold_indices[i])
        train_idx = np.array(list(all_indices - set(test_idx)))
        print(f"   Fold {i+1} Split: {len(train_idx)} train, {len(test_idx)} test")
        splits.append((train_idx, test_idx))

    return splits

print("✅ Advanced CV splitting functions are ready.")

def plot_regression(y_true, y_pred, metrics, title):
    """
    Generates and displays a regression plot for a given fold.
    """
    plt.figure(figsize=(8, 8))

    # Scatter plot
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, edgecolor='k', s=50)

    # Identity line (y=x)
    max_val = max(np.max(y_true), np.max(y_pred))
    min_val = min(np.min(y_true), np.min(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Identity Line (y=x)')

    # Display metrics on the plot
    metrics_text = (
        f"RMSE: {metrics['rmse']:.4f}\n"
        f"Pearson: {metrics['pearson']:.4f}\n"
        f"Spearman: {metrics['spearman']:.4f}\n"
        f"CI: {metrics['ci']:.4f}"
    )
    plt.text(0.05, 0.95, metrics_text, transform=plt.gca().transAxes,
             fontsize=12, verticalalignment='top',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))

    plt.title(title, fontsize=16)
    plt.xlabel("Actual pKd", fontsize=14)
    plt.ylabel("Predicted pKd", fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

print("✅ Regression plotting function is ready.")

# ==============================================================================
# 7. COMPREHENSIVE CROSS-VALIDATION & BENCHMARKING (WITH AGGREGATED PLOTTING)
# ==============================================================================
print("\n" + "="*80)
print("... 7. STARTING COMPREHENSIVE CV BENCHMARKING ...")

# --- HYPERPARAMETERS & CONFIG ---
strategies_to_run = ['GroupKFold']
# WARNING: 'SequenceSimilarity' can be very slow.
# For a quick test, use: strategies_to_run = ['KFold', 'GroupKFold']

N_SPLITS = 5
EPOCHS = 30
BATCH_SIZE = 32
EARLY_STOPPING_PATIENCE = 15    # stop if no RMSE improvement for 10 epochs
EARLY_STOPPING_MIN_DELTA = 1e-4 # minimum improvement in RMSE to reset patience
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 512
DROPOUT = 0.1

# --- MASTER RESULTS DICTIONARY ---
all_strategy_results = {}

# --- OUTER LOOP: Iterate over each CV strategy ---
for strategy in strategies_to_run:

    # --- Setup based on chosen CV Strategy ---
    if strategy == 'GroupKFold':
        print(f"\n🚀 Starting {N_SPLITS}-Fold **GroupKFold** Cross-Validation ...")
        splits = get_groupkfold_splits(df, N_SPLITS)
    elif strategy == 'SequenceSimilarity':
        print(f"\n🚀 Starting {N_SPLITS}-Fold **Sequence Similarity** Cross-Validation ...")
        splits = get_sequence_similarity_splits(df, N_SPLITS)
    else: # Default to KFold
        print(f"\n🚀 Starting {N_SPLITS}-Fold **Standard KFold** Cross-Validation ...")
        kfold = KFold(n_splits=N_SPLITS, shuffle=True, random_state=42)
        splits = kfold.split(df)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # --- Initialize lists to store results for the current strategy ---
    fold_results = [] # For statistical summary (mean, std)
    all_folds_y_true = [] # To collect all true labels for the aggregate plot
    all_folds_y_pred = [] # To collect all predictions for the aggregate plot

    PKD_LOWER_BOUND = df['Y'].min()
    PKD_UPPER_BOUND = df['Y'].max()
    PKD_BOUNDS = (PKD_LOWER_BOUND, PKD_UPPER_BOUND)
    print(f"Global pKd bounds for scaling: [{PKD_LOWER_BOUND:.2f}, {PKD_UPPER_BOUND:.2f}]")

    # --- INNER LOOP: Iterate over the folds for the current strategy ---
    for fold, (train_ids, val_ids) in enumerate(splits):
        fold_num = fold + 1
        print(f"\n===== STRATEGY: {strategy} | FOLD {fold_num}/{N_SPLITS} =====")

        train_df, val_df = df.iloc[train_ids], df.iloc[val_ids]
        train_dataset = ProteinPairEmbeddingDataset(train_df, EMBEDDING_CACHE, PKD_BOUNDS)
        val_dataset = ProteinPairEmbeddingDataset(val_df, EMBEDDING_CACHE, PKD_BOUNDS)
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn_embeddings)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE*2, shuffle=False, collate_fn=collate_fn_embeddings)

        model = BALMForEmbeddings(
            embedding_size=embedding_size,
            projected_size=PROJECTED_SIZE,
            projected_dropout=DROPOUT
        ).to(device)
        optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)

        # Training loop with Early Stopping (monitors validation RMSE)
        best_rmse = float("inf")
        best_state = None
        patience_counter = 0

        for epoch in range(EPOCHS):
            model.train()
            progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}", leave=False)
            for batch in progress_bar:
                optimizer.zero_grad()
                batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
                outputs = model(batch_gpu)
                loss = outputs['loss']
                loss.backward()
                optimizer.step()
                progress_bar.set_postfix({'train_loss': f"{loss.item():.4f}"})

            # ---- evaluate on validation at end of each epoch ----
            val_metrics, _, _, _, _ = evaluate_model(model, val_loader, device, PKD_BOUNDS)

            current_rmse = val_metrics['rmse'] if val_metrics['rmse'] is not None else float("inf")

            print(f"Epoch {epoch+1}/{EPOCHS} | Val RMSE: {current_rmse:.4f} | Best RMSE: {best_rmse:.4f} | Patience: {patience_counter}/{EARLY_STOPPING_PATIENCE}")

            # ---- early stopping check (lower RMSE is better) ----
            if (best_rmse - current_rmse) > EARLY_STOPPING_MIN_DELTA:
                best_rmse = current_rmse
                best_state = deepcopy(model.state_dict())
                patience_counter = 0
            else:
                patience_counter += 1
                if patience_counter >= EARLY_STOPPING_PATIENCE:
                    print(f"⏹️ Early stopping at epoch {epoch+1}: no RMSE improvement > {EARLY_STOPPING_MIN_DELTA} for {EARLY_STOPPING_PATIENCE} epochs.")
                    break

        # restore best weights before final fold evaluation
        if best_state is not None:
            model.load_state_dict(best_state)

        val_metrics, yt, yp, fold_pdbs, fold_sources = evaluate_model(model, val_loader, device, PKD_BOUNDS)
        
        pd.DataFrame({
            'label': yt, 
            'prediction': yp, 
            'residual': yt - yp,
            'PDB': fold_pdbs, 
            'Source Data Set': fold_sources
        }).to_csv(f"fold_{fold_num}_predictions.csv", index=False)

        print(f"Fold {fold_num} training complete (best Val RMSE: {best_rmse:.4f}).")


        # Evaluate and collect results
        # ---- Evaluate on validation set for the fold and collect results ----
        val_metrics, y_true, y_pred, fold_pdbs, fold_sources = evaluate_model(model, val_loader, device, PKD_BOUNDS)
        # Ensure arrays are 1D numpy arrays (safety)
        y_true = np.array(y_true).ravel() if y_true is not None else np.array([])
        y_pred = np.array(y_pred).ravel() if y_pred is not None else np.array([])

        print(f"Fold {fold_num} Validation Metrics:")
        for key, value in val_metrics.items():
            # Guard printing if metric is NaN
            print(f"  - {key.upper()}: {value:.4f}" if (value == value) else f"  - {key.upper()}: NaN")

        # Debug: how many validation examples this fold produced
        print(f"  ➤ Fold {fold_num} validation samples: {len(y_true)}")

        # Append fold metrics even if predictions empty (so we keep fold-level tracking),
        # but append predictions only if non-empty to avoid later aggregation errors.
        fold_results.append(val_metrics)
        if len(y_true) > 0 and len(y_pred) > 0:
            all_folds_y_true.extend(y_true.tolist())
            all_folds_y_pred.extend(y_pred.tolist())
        else:
            print(f"  ⚠️ No validation predictions to aggregate for fold {fold_num}. Skipping aggregation for this fold.")


    # --- AFTER ALL FOLDS for the strategy are done, create the AGGREGATE PLOT ---
    print(f"\n📈 Generating combined regression plot for strategy: {strategy}")

    # Convert lists to numpy arrays
    combined_y_true = np.array(all_folds_y_true)
    combined_y_pred = np.array(all_folds_y_pred)

    # Calculate overall metrics on the combined out-of-sample predictions
    overall_metrics = {
        'rmse': np.sqrt(mean_squared_error(combined_y_true, combined_y_pred)),
        'pearson': pearsonr(combined_y_true, combined_y_pred)[0],
        'spearman': spearmanr(combined_y_true, combined_y_pred)[0],
        'ci': concordance_index(combined_y_true, combined_y_pred)
    }

    # Generate the single plot for the entire strategy
    plot_regression(
        combined_y_true,
        combined_y_pred,
        overall_metrics,
        title=f"Overall Regression for {strategy} ({N_SPLITS}-Fold CV)"
    )

    # Store the per-fold results for the final statistical comparison plot
    all_strategy_results[strategy] = pd.DataFrame(fold_results)
    print(f"\n✅ Completed all folds for strategy: {strategy}")


# ==============================================================================
# 8. FINAL RESULTS COMPARISON
# ==============================================================================
print("\n" + "="*80)
print("... 8. FINAL CROSS-VALIDATION RESULTS COMPARISON ...")

# --- First, print summary tables for each strategy ---
for strategy, results_df in all_strategy_results.items():
    print(f"\n--- Summary for Strategy: {strategy} ---")
    summary_df = pd.DataFrame({
        'Mean': results_df.mean(),
        'Std Dev': results_df.std()
    })
    print(summary_df)
    print("-" * 40)

# --- Now, create the comparison error bar plots ---

def plot_comparison_results(all_results):
    """
    Generates bar plots with error bars to compare CV strategies.
    """
    metrics_to_plot = ['rmse', 'pearson', 'spearman', 'ci']

    for metric in metrics_to_plot:
        plt.figure(figsize=(10, 6))

        strategy_names = list(all_results.keys())
        mean_values = [results_df[metric].mean() for results_df in all_results.values()]
        std_devs = [results_df[metric].std() for results_df in all_results.values()]

        bars = plt.bar(strategy_names, mean_values, yerr=std_devs, capsize=5,
                       color=sns.color_palette("viridis", len(strategy_names)), alpha=0.8)

        plt.ylabel(metric.upper(), fontsize=14)
        plt.title(f"Comparison of CV Strategies for {metric.upper()}", fontsize=16)
        plt.xticks(rotation=0, ha="center", fontsize=12)

        # Add values on top of bars
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.3f}', va='bottom' if yval >= 0 else 'top', ha='center')

        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

if not all_strategy_results:
    print("⚠️ No results were generated. Skipping final comparison plot.")
else:
    plot_comparison_results(all_strategy_results)

print("\n✅ Benchmark analysis complete.")
```

```python
# ==============================================================================
# 1. IMPORTS AND INITIAL SETUP
# ==============================================================================
import os
import sys
import time
import json
import random
import pickle
from pathlib import Path
from datetime import datetime
from copy import deepcopy

# Standard data science libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import warnings

# Machine learning libraries
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr, spearmanr

# PyTorch and Transformers
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import AutoModelForCausalLM, AutoTokenizer, get_linear_schedule_with_warmup

# Set style for plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
warnings.filterwarnings('ignore', category=UserWarning)

print("🧬 BALM Adaptation for ProGen-2 Cached Embeddings")
print("⚡ Fast Training with 5-Fold Cross-Validation")
print("="*80)

def setup_reproducibility(seed=42):
    """Setup complete reproducibility."""
    print(f"🔒 Setting up reproducibility with seed {seed}")
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    os.environ['PYTHONHASHSEED'] = str(seed)
    print(f"✅ Reproducibility setup complete")

setup_reproducibility(42)

# ==============================================================================
# 2. PROGEN-2 EMBEDDING EXTRACTOR (CORRECTED WITH CACHE METHODS)
# ==============================================================================

class FastEmbeddingExtractor:
    def __init__(self, model_name="hugohrban/progen2-medium", max_seq_len=1024, batch_size=16, device="auto"):
        self.model_name, self.max_seq_len, self.batch_size = model_name, max_seq_len, batch_size
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu") if device == "auto" else torch.device(device)
        self.dtype = torch.float16 if self.device.type == 'cuda' else torch.float32

        print(f"🔧 Initializing Extractor on {self.device}")
        
        # Standard way to load with dtype
        self.model = AutoModelForCausalLM.from_pretrained(
            model_name, 
            trust_remote_code=True, 
            torch_dtype=self.dtype 
        ).to(self.device).eval()
        
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token
        self.tokenizer.padding_side = 'left'

        self.embedding_size = self.model.config.n_embd
        self.embedding_cache = {}

    def compute_embeddings_batch(self, sequences):
        if not sequences: return torch.empty(0, self.embedding_size)

        # CRITICAL: Replace '|' with two EOS tokens for multi-chain support
        sep = self.tokenizer.eos_token
        processed_seqs = [s.replace('|', f"{sep}{sep}") for s in sequences]

        inputs = self.tokenizer(processed_seqs, padding=True, truncation=True, max_length=self.max_seq_len, return_tensors="pt")
        inputs = {k: v.to(self.device) for k, v in inputs.items()}

        with torch.no_grad():
            outputs = self.model(**inputs, output_hidden_states=True)
            last_hidden_states = outputs.hidden_states[-1] # [Batch, Seq, Dim]
            
            # CRITICAL: Mean Pooling (required for concatenated sequences)
            mask = inputs['attention_mask'].unsqueeze(-1).expand(last_hidden_states.size()).float()
            sum_embeddings = torch.sum(last_hidden_states * mask, 1)
            sum_mask = torch.clamp(mask.sum(1), min=1e-9)
            embeddings = (sum_embeddings / sum_mask).cpu()

        return embeddings

    def get_embeddings(self, sequences, use_cache=True, show_progress=True):
        if isinstance(sequences, str): sequences = [sequences]
        embeddings, to_compute, to_compute_indices = [], [], []
        for i, seq in enumerate(sequences):
            if use_cache and seq in self.embedding_cache:
                embeddings.append(self.embedding_cache[seq])
            else:
                embeddings.append(None)
                to_compute.append(seq)
                to_compute_indices.append(i)
        if to_compute:
            if show_progress: print(f"🧮 Computing {len(to_compute)} new embeddings...")
            computed_embeddings = []
            pbar = tqdm(range(0, len(to_compute), self.batch_size), desc="Computing embeddings", disable=not show_progress)
            for start_idx in pbar:
                end_idx = min(start_idx + self.batch_size, len(to_compute))
                batch_sequences = to_compute[start_idx:end_idx]
                batch_embeddings = self.compute_embeddings_batch(batch_sequences)
                computed_embeddings.append(batch_embeddings)
                if use_cache:
                    for j, seq in enumerate(batch_sequences):
                        self.embedding_cache[seq] = batch_embeddings[j:j+1]
                if torch.cuda.is_available():
                    torch.cuda.empty_cache()
            if computed_embeddings:
                all_computed = torch.cat(computed_embeddings, dim=0)
                for i, orig_idx in enumerate(to_compute_indices):
                    embeddings[orig_idx] = all_computed[i:i+1]

        if not embeddings: return torch.empty(0, self.embedding_size)
        final_embeddings = torch.cat(embeddings, dim=0)
        if show_progress: print(f"✅ Generated embeddings: {final_embeddings.shape}")
        return final_embeddings

    # --- ADDED THESE MISSING METHODS ---
    def save_cache(self, cache_path):
        """Saves the current embedding cache to a file using pickle."""
        print(f"💾 Saving embedding cache to {cache_path}")
        with open(cache_path, 'wb') as f:
            pickle.dump(self.embedding_cache, f)
        print(f"✅ Saved {len(self.embedding_cache)} cached embeddings.")

    def load_cache(self, cache_path):
        """Loads an embedding cache from a file if it exists."""
        if os.path.exists(cache_path):
            print(f"📥 Loading embedding cache from {cache_path}")
            try:
                with open(cache_path, 'rb') as f:
                    self.embedding_cache = pickle.load(f)
                print(f"✅ Loaded {len(self.embedding_cache)} cached embeddings.")
                return True
            except (pickle.UnpicklingError, EOFError) as e:
                print(f"⚠️ Could not load cache file (it might be corrupted): {e}. A new cache will be created.")
                self.embedding_cache = {}
                return False
        return False
    # ------------------------------------

# ==============================================================================
# 3. LOAD YOUR DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 3. LOADING DATA FROM Data.csv ...")

protein1_col = 'Target'    
protein2_col = 'proteina'  
target_col = 'Y'         
pdb_col_raw = 'PDB'
source_dataset_col = 'Source Data Set' # NEW

try:
    df_raw = pd.read_csv(r'/teamspace/studios/this_studio/PPB_Affinity_Sequences_Final (version 1).csv')
    print(f"Successfully loaded Data.csv with {len(df_raw)} rows.")

    # Include Source Data Set in the selection
    df = df_raw[[protein1_col, protein2_col, target_col, pdb_col_raw, source_dataset_col]].copy()
    
    df.rename(columns={
        protein1_col: 'Target',
        protein2_col: 'proteina',
        target_col: 'Y',
        pdb_col_raw: 'PDB',
        source_dataset_col: 'SourceDataSet' # Rename for easier access
    }, inplace=True)

    # --- Data Validation ---
    missing_values = df.isnull().sum()
    if missing_values.sum() > 0:
        print(f"\nWarning: Found {missing_values.sum()} total missing values. Dropping rows with NaNs.")
        # FIX: Added 'subset=' keyword here
        df.dropna(subset=['Target', 'proteina', 'Y', 'PDB'], inplace=True)
        print(f"Dataset now has {len(df)} rows after cleaning.")

    # Ensure data types are correct
    df['Target'] = df['Target'].astype(str)
    df['proteina'] = df['proteina'].astype(str)
    df['SourceDataSet'] = df['SourceDataSet'].astype(str) # NEW
    df['Y'] = pd.to_numeric(df['Y'], errors='coerce') 
    df['PDB'] = df['PDB'].astype(str)
    df.dropna(subset=['Y'], inplace=True) 

    print(f"\n✅ Data is loaded and cleaned. Final dataset size: {len(df)} samples.")

except FileNotFoundError:
    print("❌ ERROR: 'Data.csv' not found.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB', 'SourceDataSet'])
except KeyError as e:
    print(f"❌ ERROR: Column {e} not found in 'Data.csv'.")
    df = pd.DataFrame(columns=['Target', 'proteina', 'Y', 'PDB', 'SourceDataSet'])

# ==============================================================================
# 4. COMPUTE OR LOAD CACHED EMBEDDINGS
# ==============================================================================
print("\n" + "="*80)
print("... 4. MANAGING PROGEN-2 EMBEDDINGS CACHE ...")

# --- Define the path for your persistent embedding cache ---
CACHE_FILE = "progen2_embeddings_cache.pkl"

# Instantiate the extractor
# This is needed in both cases to get the model's embedding size and to hold the cache
extractor = FastEmbeddingExtractor(batch_size=32)

# Try to load the cache from the file
cache_loaded = extractor.load_cache(CACHE_FILE)

# Collect all unique sequences from the dataframe. This is needed to check if
# the loaded cache is sufficient or if we need to compute more embeddings.
unique_sequences = pd.concat([df['Target'], df['proteina']]).unique().tolist()
print(f"Found {len(unique_sequences)} unique protein sequences in the dataset.")

# Check if any sequences are missing from the loaded cache
if cache_loaded:
    cached_keys = set(extractor.embedding_cache.keys())
    missing_sequences = [seq for seq in unique_sequences if seq not in cached_keys]
    if not missing_sequences:
        print("✅ All required embeddings were successfully loaded from the cache file.")
    else:
        print(f"⚠️ Cache is missing {len(missing_sequences)} embeddings. They will be computed now.")
else:
    print(f"Cache file '{CACHE_FILE}' not found. All embeddings will be computed.")

# This single call will now use the loaded cache and only compute what's missing.
# If the cache was empty or didn't exist, it computes everything.
_ = extractor.get_embeddings(unique_sequences, show_progress=True)

# Save the updated cache back to the file if any new embeddings were computed.
# This ensures the cache file is always up-to-date with the current dataset.
if not cache_loaded or (locals().get('missing_sequences') and missing_sequences):
    extractor.save_cache(CACHE_FILE)

# The EMBEDDING_CACHE is now our master lookup table for the training process
EMBEDDING_CACHE = extractor.embedding_cache
embedding_size = extractor.embedding_size
print(f"✅ Embedding cache is ready with {len(EMBEDDING_CACHE)} entries.")
print(f"Embedding dimension: {embedding_size}")

# ==============================================================================
# 5. ADAPTED MODEL AND DATASET
# ==============================================================================
print("\n" + "="*80)
print("... 5. DEFINING ADAPTED MODEL AND DATASET ...")

class BALMForEmbeddings(nn.Module):
    """
    Adapted BALM model that accepts pre-computed ProGen-2 embeddings.
    This version is self-contained and does not need base_model.py or utils.py.
    """
    def __init__(
        self,
        embedding_size=1024, # ProGen-2 Small
        projected_size=512,
        projected_dropout=0.1,
        relu_before_cosine=False
    ):
        super().__init__()
        self.embedding_size = embedding_size

        self.protein_projection = nn.Linear(embedding_size, projected_size)
        self.proteina_projection = nn.Linear(embedding_size, projected_size)
        self.dropout = nn.Dropout(projected_dropout)
        self.relu_before_cosine = relu_before_cosine
        self.loss_fn = nn.MSELoss()

        print("✅ BALMForEmbeddings model initialized.")

    def forward(self, batch_input):
        # This model expects embeddings directly, not sequences
        protein_embedding = batch_input["protein_embedding"]
        proteina_embedding = batch_input["proteina_embedding"]

        # Apply dropout for regularization
        protein_embedding = self.dropout(protein_embedding)
        proteina_embedding = self.dropout(proteina_embedding)

        # Project to shared latent space
        protein_projected = self.protein_projection(protein_embedding)
        proteina_projected = self.proteina_projection(proteina_embedding)

        if self.relu_before_cosine:
            protein_projected = F.relu(protein_projected)
            proteina_projected = F.relu(proteina_projected)

        # L2 normalize before cosine similarity for numerical stability
        protein_projected = F.normalize(protein_projected, p=2, dim=1)
        proteina_projected = F.normalize(proteina_projected, p=2, dim=1)

        # Compute cosine similarity
        cosine_similarity = F.cosine_similarity(protein_projected, proteina_projected)
        cosine_similarity = torch.clamp(cosine_similarity, -0.9999, 0.9999)

        forward_output = {"cosine_similarity": cosine_similarity}

        # Compute loss if labels are provided
        if "labels" in batch_input and batch_input["labels"] is not None:
            labels = batch_input["labels"]
            loss = self.loss_fn(cosine_similarity, labels)
            forward_output["loss"] = loss

        return forward_output

class ProteinPairEmbeddingDataset(Dataset):
    def __init__(self, dataframe, embedding_cache, pkd_bounds):
        self.data = dataframe.reset_index(drop=True)
        self.embedding_cache = embedding_cache
        self.pkd_lower, self.pkd_upper = pkd_bounds
        self.pkd_range = self.pkd_upper - self.pkd_lower

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        row = self.data.iloc[idx]
        p1_emb = self.embedding_cache[row["Target"]].squeeze(0)
        p2_emb = self.embedding_cache[row["proteina"]].squeeze(0)
        scaled_label = ((float(row['Y']) - self.pkd_lower) / self.pkd_range) * 2 - 1

        return {
            "protein_embedding": p1_emb.float(),
            "proteina_embedding": p2_emb.float(),
            "labels": torch.tensor(scaled_label, dtype=torch.float32),
            "original_pkds": torch.tensor(row['Y'], dtype=torch.float32),
            "pdbs": row["PDB"],
            "sources": row["SourceDataSet"] # Matches Section 3 renaming
        }

def collate_fn_embeddings(batch):
    return {
        "protein_embedding": torch.stack([item['protein_embedding'] for item in batch]),
        "proteina_embedding": torch.stack([item['proteina_embedding'] for item in batch]),
        "labels": torch.stack([item['labels'] for item in batch]),
        "original_pkds": torch.stack([item['original_pkds'] for item in batch]),
        "pdbs": [item['pdbs'] for item in batch],
        "sources": [item['sources'] for item in batch]
    }


# ==============================================================================
# 6. METRICS AND EVALUATION FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6. DEFINING METRICS AND EVALUATION ...")

def concordance_index(y_true, y_pred):
    """Calculate concordance index."""
    try:
        y_true, y_pred = np.array(y_true), np.array(y_pred)
        n = len(y_true)
        if n < 2: return 1.0 # Not enough pairs to compare

        concordant = 0
        discordant = 0
        tied_pred = 0

        for i in range(n):
            for j in range(i + 1, n):
                if y_true[i] == y_true[j]:
                    continue

                if y_pred[i] == y_pred[j]:
                    tied_pred += 1
                    continue

                if (y_pred[i] > y_pred[j] and y_true[i] > y_true[j]) or \
                   (y_pred[i] < y_pred[j] and y_true[i] < y_true[j]):
                    concordant += 1
                else:
                    discordant += 1

        if (concordant + discordant + tied_pred) == 0:
            return 1.0

        ci = (concordant + 0.5 * tied_pred) / (concordant + discordant + tied_pred)
        return ci
    except Exception:
        return np.nan

def evaluate_model(model, data_loader, device, pkd_bounds):
    """
    Evaluate the model.
    RETURNS: A tuple of (metrics_dict, actual_labels_array, predicted_labels_array, pdbs, sources)
    """
    model.eval()
    all_labels, all_preds, all_pdbs, all_srcs = [], [], [], []

    pkd_lower, pkd_upper = pkd_bounds
    pkd_range = pkd_upper - pkd_lower

    with torch.no_grad():
        for batch in data_loader:
            batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
            outputs = model(batch_gpu)
            cosine_sim = outputs['cosine_similarity'].cpu()
            
            # De-scale prediction back to pKd
            preds_pkd = ((cosine_sim + 1) / 2) * pkd_range + pkd_lower

            all_labels.extend(batch['original_pkds'].numpy())
            all_preds.extend(preds_pkd.numpy())
            all_pdbs.extend(batch['pdbs'])    
            all_srcs.extend(batch['sources']) 

    labels = np.array(all_labels)
    preds = np.array(all_preds)

    # Clean out any NaNs just in case
    mask = ~np.isnan(labels) & ~np.isnan(preds)
    clean_labels, clean_preds = labels[mask], preds[mask]

    if len(clean_labels) < 2:
        metrics = {"rmse": np.nan, "pearson": np.nan, "spearman": np.nan, "ci": np.nan}
    else:
        metrics = {
            'rmse': np.sqrt(mean_squared_error(clean_labels, clean_preds)),
            'pearson': pearsonr(clean_labels, clean_preds)[0],
            'spearman': spearmanr(clean_labels, clean_preds)[0],
            'ci': concordance_index(clean_labels, clean_preds)
        }

    return metrics, labels, preds, all_pdbs, all_srcs
# ==============================================================================
# 6.A: EXPERIMENT CONFIGURATION
# ==============================================================================
print("\n" + "="*80)
print("... 6.A: CONFIGURING THE EXPERIMENT ...")

# --- CHOOSE YOUR CROSS-VALIDATION STRATEGY ---
# Options:
# 'KFold'              : Standard random split. Fastest, but risks data leakage.
# 'GroupKFold'         : Cold-target split. Ensures a protein seen in the test set was never seen in the train set.
# 'SequenceSimilarity' : Cold-cluster split. Groups similar proteins and puts them in the same fold. Most rigorous, but VERY SLOW.

CV_STRATEGY = 'GroupKFold' # <-- CHANGE THIS TO 'KFold' or 'SequenceSimilarity'

# --- HYPERPARAMETERS ---
N_SPLITS = 5
EPOCHS = 30 # Increase for better performance, e.g., 50-100
BATCH_SIZE = 32
EARLY_STOPPING_PATIENCE = 15    # stop if no RMSE improvement for 10 epochs
EARLY_STOPPING_MIN_DELTA = 1e-4 # minimum improvement in RMSE to reset patience
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 1024
DROPOUT = 0.1

print(f"✅ Configuration set. Using CV Strategy: {CV_STRATEGY}")

# ==============================================================================
# 6.B: ADVANCED CROSS-VALIDATION SPLITTING FUNCTIONS
# ==============================================================================
print("\n" + "="*80)
print("... 6.B: DEFINING ADVANCED CV SPLITTING LOGIC ...")

# Required imports for the new strategies
from sklearn.model_selection import GroupKFold
from Bio.Align import PairwiseAligner
from scipy.cluster.hierarchy import linkage, fcluster
from multiprocessing import Pool, cpu_count
from typing import List, Tuple

# --- Strategy 1: GroupKFold (Cold Target) ---

def get_groupkfold_splits(df: pd.DataFrame, n_splits: int):
    """
    Generates train/test splits using GroupKFold to ensure cold-target validation.
    Proteins in the 'Target' column are used as groups.
    """
    print(f"\n❄️ Preparing {n_splits}-Fold **GroupKFold (Cold Target)** splits...")
    grouping_column = 'PDB'
    groups = df[grouping_column].factorize()[0]
    gkf = GroupKFold(n_splits=n_splits)

    for fold, (train_idx, test_idx) in enumerate(gkf.split(df, groups=groups)):
        # Verification step
        train_targets = set(df.iloc[train_idx][grouping_column])
        test_targets = set(df.iloc[test_idx][grouping_column])
        overlap = train_targets.intersection(test_targets)

        if len(overlap) > 0:
            print(f"❌ Fold {fold+1} FAILED: Target leakage detected with {len(overlap)} overlapping proteins!")
            # In a strict setting, you might raise an error here. For now, we'll just warn.
        else:
            print(f"  ✅ Fold {fold+1}: No target leakage detected.")

        yield train_idx, test_idx

# --- Strategy 2: Sequence Similarity (Cold Cluster) ---

def compute_pairwise_similarity(args: Tuple[str, str, int, int]) -> Tuple[int, int, float]:
    """Helper for parallel similarity calculation."""
    seq1, seq2, i, j = args
    aligner = PairwiseAligner()
    aligner.mode = 'global'
    aligner.match_score = 1
    aligner.mismatch_score = -1
    aligner.open_gap_score = -1
    aligner.extend_gap_score = -0.5
    try:
        score = aligner.score(seq1, seq2)
        sim = score / max(len(seq1), len(seq2), 1)
    except Exception:
        sim = 0.0
    return i, j, sim

def get_sequence_similarity_splits(df: pd.DataFrame, n_splits: int, identity_threshold: float = 0.5):
    """
    Generates train/test splits based on sequence similarity clustering.
    This is computationally intensive.
    """
    print(f"\n🧬 Preparing {n_splits}-Fold **Sequence Similarity** splits (Threshold={identity_threshold})...")
    print("⚠️ This process can be very slow, especially the first time.")

    # 1. Extract unique sequences
    print("   Step 1/5: Extracting unique sequences...")
    unique_sequences = list(pd.concat([df['Target'], df['proteina']]).unique())
    seq_to_idx = {seq: i for i, seq in enumerate(unique_sequences)}
    n_seqs = len(unique_sequences)

    # 2. Compute similarity matrix
    print(f"   Step 2/5: Computing similarity matrix for {n_seqs} sequences (parallelized)...")
    similarity_matrix = np.identity(n_seqs)
    pairs = [(unique_sequences[i], unique_sequences[j], i, j) for i in range(n_seqs) for j in range(i + 1, n_seqs)]

    with Pool(processes=max(1, cpu_count() - 1)) as pool:
        results = list(tqdm(pool.imap(compute_pairwise_similarity, pairs), total=len(pairs), desc="Calculating Similarities"))

    for i, j, sim in results:
        similarity_matrix[i, j] = similarity_matrix[j, i] = sim

    # 3. Cluster sequences
    print("   Step 3/5: Clustering sequences...")
    distance_matrix = 1 - similarity_matrix
    # Ensure distances are non-negative
    distance_matrix[distance_matrix < 0] = 0
    condensed_distance_matrix = distance_matrix[np.triu_indices(n_seqs, k=1)]

    Z = linkage(condensed_distance_matrix, method='complete')
    clusters = fcluster(Z, t=(1 - identity_threshold), criterion='distance')
    seq_to_cluster_id = {seq: cluster_id for seq, cluster_id in zip(unique_sequences, clusters)}
    num_clusters = len(set(clusters))
    print(f"   Found {num_clusters} clusters.")

    # 4. Assign clusters to folds
    print("   Step 4/5: Assigning clusters to folds...")
    cluster_ids = np.unique(clusters)
    np.random.shuffle(cluster_ids) # Randomize cluster order before assigning

    # Assign each cluster ID to a fold
    cluster_to_fold = {cid: i % n_splits for i, cid in enumerate(cluster_ids)}

    # 5. Create final splits
    print("   Step 5/5: Generating final train/test indices...")
    fold_indices = [[] for _ in range(n_splits)]
    for i, row in tqdm(df.iterrows(), total=len(df), desc="Assigning Pairs to Folds"):
        target_cid = seq_to_cluster_id.get(row['Target'])
        proteina_cid = seq_to_cluster_id.get(row['proteina'])

        # A pair belongs to a fold if EITHER of its proteins' clusters belongs to that fold
        target_fold = cluster_to_fold.get(target_cid, -1)
        proteina_fold = cluster_to_fold.get(proteina_cid, -1)

        # If proteins are in different clusters assigned to different folds,
        # we can decide on a rule. Let's assign to the target's fold.
        # This is a simplification; more complex rules could be used.
        assigned_fold = target_fold if target_fold != -1 else proteina_fold

        if assigned_fold != -1:
            fold_indices[assigned_fold].append(i)

    splits = []
    all_indices = set(df.index)
    for i in range(n_splits):
        test_idx = np.array(fold_indices[i])
        train_idx = np.array(list(all_indices - set(test_idx)))
        print(f"   Fold {i+1} Split: {len(train_idx)} train, {len(test_idx)} test")
        splits.append((train_idx, test_idx))

    return splits

print("✅ Advanced CV splitting functions are ready.")

def plot_regression(y_true, y_pred, metrics, title):
    """
    Generates and displays a regression plot for a given fold.
    """
    plt.figure(figsize=(8, 8))

    # Scatter plot
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, edgecolor='k', s=50)

    # Identity line (y=x)
    max_val = max(np.max(y_true), np.max(y_pred))
    min_val = min(np.min(y_true), np.min(y_pred))
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Identity Line (y=x)')

    # Display metrics on the plot
    metrics_text = (
        f"RMSE: {metrics['rmse']:.4f}\n"
        f"Pearson: {metrics['pearson']:.4f}\n"
        f"Spearman: {metrics['spearman']:.4f}\n"
        f"CI: {metrics['ci']:.4f}"
    )
    plt.text(0.05, 0.95, metrics_text, transform=plt.gca().transAxes,
             fontsize=12, verticalalignment='top',
             bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5))

    plt.title(title, fontsize=16)
    plt.xlabel("Actual pKd", fontsize=14)
    plt.ylabel("Predicted pKd", fontsize=14)
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

print("✅ Regression plotting function is ready.")

# ==============================================================================
# 7. COMPREHENSIVE CROSS-VALIDATION & BENCHMARKING (WITH AGGREGATED PLOTTING)
# ==============================================================================
print("\n" + "="*80)
print("... 7. STARTING COMPREHENSIVE CV BENCHMARKING ...")

# --- HYPERPARAMETERS & CONFIG ---
strategies_to_run = ['GroupKFold']
# WARNING: 'SequenceSimilarity' can be very slow.
# For a quick test, use: strategies_to_run = ['KFold', 'GroupKFold']

N_SPLITS = 5
EPOCHS = 30
BATCH_SIZE = 32
EARLY_STOPPING_PATIENCE = 15    # stop if no RMSE improvement for 10 epochs
EARLY_STOPPING_MIN_DELTA = 1e-4 # minimum improvement in RMSE to reset patience
LEARNING_RATE = 1e-4
PROJECTED_SIZE = 1024
DROPOUT = 0.1

# --- MASTER RESULTS DICTIONARY ---
all_strategy_results = {}

# --- OUTER LOOP: Iterate over each CV strategy ---
for strategy in strategies_to_run:

    # --- Setup based on chosen CV Strategy ---
    if strategy == 'GroupKFold':
        print(f"\n🚀 Starting {N_SPLITS}-Fold **GroupKFold** Cross-Validation ...")
        splits = get_groupkfold_splits(df, N_SPLITS)
    elif strategy == 'SequenceSimilarity':
        print(f"\n🚀 Starting {N_SPLITS}-Fold **Sequence Similarity** Cross-Validation ...")
        splits = get_sequence_similarity_splits(df, N_SPLITS)
    else: # Default to KFold
        print(f"\n🚀 Starting {N_SPLITS}-Fold **Standard KFold** Cross-Validation ...")
        kfold = KFold(n_splits=N_SPLITS, shuffle=True, random_state=42)
        splits = kfold.split(df)

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # --- Initialize lists to store results for the current strategy ---
    fold_results = [] # For statistical summary (mean, std)
    all_folds_y_true = [] # To collect all true labels for the aggregate plot
    all_folds_y_pred = [] # To collect all predictions for the aggregate plot

    PKD_LOWER_BOUND = df['Y'].min()
    PKD_UPPER_BOUND = df['Y'].max()
    PKD_BOUNDS = (PKD_LOWER_BOUND, PKD_UPPER_BOUND)
    print(f"Global pKd bounds for scaling: [{PKD_LOWER_BOUND:.2f}, {PKD_UPPER_BOUND:.2f}]")

    # --- INNER LOOP: Iterate over the folds for the current strategy ---
    for fold, (train_ids, val_ids) in enumerate(splits):
        fold_num = fold + 1
        print(f"\n===== STRATEGY: {strategy} | FOLD {fold_num}/{N_SPLITS} =====")

        train_df, val_df = df.iloc[train_ids], df.iloc[val_ids]
        train_dataset = ProteinPairEmbeddingDataset(train_df, EMBEDDING_CACHE, PKD_BOUNDS)
        val_dataset = ProteinPairEmbeddingDataset(val_df, EMBEDDING_CACHE, PKD_BOUNDS)
        train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn_embeddings)
        val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE*2, shuffle=False, collate_fn=collate_fn_embeddings)

        model = BALMForEmbeddings(
            embedding_size=embedding_size,
            projected_size=PROJECTED_SIZE,
            projected_dropout=DROPOUT
        ).to(device)
        optimizer = AdamW(model.parameters(), lr=LEARNING_RATE)

        # Training loop with Early Stopping (monitors validation RMSE)
        best_rmse = float("inf")
        best_state = None
        patience_counter = 0

        for epoch in range(EPOCHS):
            model.train()
            progress_bar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}", leave=False)
            for batch in progress_bar:
                optimizer.zero_grad()
                batch_gpu = {k: v.to(device) if isinstance(v, torch.Tensor) else v for k, v in batch.items()}
                outputs = model(batch_gpu)
                loss = outputs['loss']
                loss.backward()
                optimizer.step()
                progress_bar.set_postfix({'train_loss': f"{loss.item():.4f}"})

            # ---- evaluate on validation at end of each epoch ----
            val_metrics, _, _, _, _ = evaluate_model(model, val_loader, device, PKD_BOUNDS)

            current_rmse = val_metrics['rmse'] if val_metrics['rmse'] is not None else float("inf")

            print(f"Epoch {epoch+1}/{EPOCHS} | Val RMSE: {current_rmse:.4f} | Best RMSE: {best_rmse:.4f} | Patience: {patience_counter}/{EARLY_STOPPING_PATIENCE}")

            # ---- early stopping check (lower RMSE is better) ----
            if (best_rmse - current_rmse) > EARLY_STOPPING_MIN_DELTA:
                best_rmse = current_rmse
                best_state = deepcopy(model.state_dict())
                patience_counter = 0
            else:
                patience_counter += 1
                if patience_counter >= EARLY_STOPPING_PATIENCE:
                    print(f"⏹️ Early stopping at epoch {epoch+1}: no RMSE improvement > {EARLY_STOPPING_MIN_DELTA} for {EARLY_STOPPING_PATIENCE} epochs.")
                    break

        # restore best weights before final fold evaluation
        if best_state is not None:
            model.load_state_dict(best_state)

        val_metrics, yt, yp, fold_pdbs, fold_sources = evaluate_model(model, val_loader, device, PKD_BOUNDS)
        
        pd.DataFrame({
            'label': yt, 
            'prediction': yp, 
            'residual': yt - yp,
            'PDB': fold_pdbs, 
            'Source Data Set': fold_sources
        }).to_csv(f"fold_{fold_num}_predictions.csv", index=False)

        print(f"Fold {fold_num} training complete (best Val RMSE: {best_rmse:.4f}).")


        # Evaluate and collect results
        # ---- Evaluate on validation set for the fold and collect results ----
        val_metrics, y_true, y_pred, fold_pdbs, fold_sources = evaluate_model(model, val_loader, device, PKD_BOUNDS)
        # Ensure arrays are 1D numpy arrays (safety)
        y_true = np.array(y_true).ravel() if y_true is not None else np.array([])
        y_pred = np.array(y_pred).ravel() if y_pred is not None else np.array([])

        print(f"Fold {fold_num} Validation Metrics:")
        for key, value in val_metrics.items():
            # Guard printing if metric is NaN
            print(f"  - {key.upper()}: {value:.4f}" if (value == value) else f"  - {key.upper()}: NaN")

        # Debug: how many validation examples this fold produced
        print(f"  ➤ Fold {fold_num} validation samples: {len(y_true)}")

        # Append fold metrics even if predictions empty (so we keep fold-level tracking),
        # but append predictions only if non-empty to avoid later aggregation errors.
        fold_results.append(val_metrics)
        if len(y_true) > 0 and len(y_pred) > 0:
            all_folds_y_true.extend(y_true.tolist())
            all_folds_y_pred.extend(y_pred.tolist())
        else:
            print(f"  ⚠️ No validation predictions to aggregate for fold {fold_num}. Skipping aggregation for this fold.")


    # --- AFTER ALL FOLDS for the strategy are done, create the AGGREGATE PLOT ---
    print(f"\n📈 Generating combined regression plot for strategy: {strategy}")

    # Convert lists to numpy arrays
    combined_y_true = np.array(all_folds_y_true)
    combined_y_pred = np.array(all_folds_y_pred)

    # Calculate overall metrics on the combined out-of-sample predictions
    overall_metrics = {
        'rmse': np.sqrt(mean_squared_error(combined_y_true, combined_y_pred)),
        'pearson': pearsonr(combined_y_true, combined_y_pred)[0],
        'spearman': spearmanr(combined_y_true, combined_y_pred)[0],
        'ci': concordance_index(combined_y_true, combined_y_pred)
    }

    # Generate the single plot for the entire strategy
    plot_regression(
        combined_y_true,
        combined_y_pred,
        overall_metrics,
        title=f"Overall Regression for {strategy} ({N_SPLITS}-Fold CV)"
    )

    # Store the per-fold results for the final statistical comparison plot
    all_strategy_results[strategy] = pd.DataFrame(fold_results)
    print(f"\n✅ Completed all folds for strategy: {strategy}")


# ==============================================================================
# 8. FINAL RESULTS COMPARISON
# ==============================================================================
print("\n" + "="*80)
print("... 8. FINAL CROSS-VALIDATION RESULTS COMPARISON ...")

# --- First, print summary tables for each strategy ---
for strategy, results_df in all_strategy_results.items():
    print(f"\n--- Summary for Strategy: {strategy} ---")
    summary_df = pd.DataFrame({
        'Mean': results_df.mean(),
        'Std Dev': results_df.std()
    })
    print(summary_df)
    print("-" * 40)

# --- Now, create the comparison error bar plots ---

def plot_comparison_results(all_results):
    """
    Generates bar plots with error bars to compare CV strategies.
    """
    metrics_to_plot = ['rmse', 'pearson', 'spearman', 'ci']

    for metric in metrics_to_plot:
        plt.figure(figsize=(10, 6))

        strategy_names = list(all_results.keys())
        mean_values = [results_df[metric].mean() for results_df in all_results.values()]
        std_devs = [results_df[metric].std() for results_df in all_results.values()]

        bars = plt.bar(strategy_names, mean_values, yerr=std_devs, capsize=5,
                       color=sns.color_palette("viridis", len(strategy_names)), alpha=0.8)

        plt.ylabel(metric.upper(), fontsize=14)
        plt.title(f"Comparison of CV Strategies for {metric.upper()}", fontsize=16)
        plt.xticks(rotation=0, ha="center", fontsize=12)

        # Add values on top of bars
        for bar in bars:
            yval = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2.0, yval, f'{yval:.3f}', va='bottom' if yval >= 0 else 'top', ha='center')

        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

if not all_strategy_results:
    print("⚠️ No results were generated. Skipping final comparison plot.")
else:
    plot_comparison_results(all_strategy_results)

print("\n✅ Benchmark analysis complete.")
```
